```python
import os
import json
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
#@from>rookid19>save:data.js
# List of URLs to scrape
urls = [
  "https://www.houseplans.com/plan/2047-square-feet-4-bedroom-3-bathroom-2-garage-country-craftsman-farmhouse-ranch-sp346871",
  "https://www.houseplans.com/plan/1642-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-modern-sp289231",
  "https://www.houseplans.com/plan/3804-square-feet-4-bedroom-3-5-bathroom-3-garage-modern-european-sp295955",
  "https://www.houseplans.com/plan/2497-square-feet-1-bedroom-1-bathroom-3-garage-farmhouse-craftsman-sp274113",
  "https://www.houseplans.com/plan/1752-square-feet-1-bedroom-1-bathroom-2-garage-farmhouse-craftsman-cottage-sp274048",
  "https://www.houseplans.com/plan/924-square-feet-2-bedroom-1-bathroom-3-garage-modern-traditional-ranch-sp296198",
  "https://www.houseplans.com/plan/896-square-feet-2-bedroom-1-bathroom-2-garage-country-sp321779",
  "https://www.houseplans.com/plan/896-square-feet-2-bedroom-1-bathroom-2-garage-country-ranch-sp328748",
  "https://www.houseplans.com/plan/890-square-feet-2-bedroom-1-bathroom-0-garage-farmhouse-southern-traditional-cottage-sp263747",
  "https://www.houseplans.com/plan/1108-square-feet-2-bedroom-1-bathroom-2-garage-barndominium-sp335918",
  "https://www.houseplans.com/plan/1160-square-feet-2-bedroom-1-bathroom-0-garage-ranch-39080",
  "https://www.houseplans.com/plan/1140-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp325825",
  "https://www.houseplans.com/plan/1945-square-feet-1-bedroom-2-5-bathroom-2-garage-country-ranch-sp314521",
  "https://www.houseplans.com/plan/1737-square-feet-1-bedroom-1-5-bathroom-2-garage-ranch-craftsman-traditional-country-sp268151",
  "https://www.houseplans.com/plan/1200-square-feet-1-bedroom-1-5-bathroom-0-garage-country-sp325830",
  "https://www.houseplans.com/plan/1824-square-feet-1-bedroom-1-5-bathroom-2-garage-country-sp301668",
  "https://www.houseplans.com/plan/525-square-feet-1-bedroom-1-bathroom-2-garage-farmhouse-southern-country-craftsman-sp299715",
  "https://www.houseplans.com/plan/3099-square-feet-1-bedroom-1-bathroom-0-garage-traditional-40579",
  "https://www.houseplans.com/plan/792-square-feet-1-bedrooms-1-bathroom-cape-cod-house-plans-2-garage-32013",
  "https://www.houseplans.com/plan/562-square-feet-1-bedroom-1-bathroom-2-garage-country-traditional-sp215050",
  "https://www.houseplans.com/plan/662-square-feet-1-bedroom-1-bathroom-3-garage-cabin-sp268101",
  "https://www.houseplans.com/plan/1960-square-feet-1-bedroom-1-bathroom-3-garage-cabin-country-colonial-traditional-sp271756",
  "https://www.houseplans.com/plan/1408-square-feet-1-bedroom-1-5-bathroom-0-garage-farmhouse-sp325814",
  "https://www.houseplans.com/plan/2235-square-feet-4-bedroom-3-bathroom-0-garage-country-traditional-farmhouse-sp324653",
  "https://www.houseplans.com/plan/2771-square-feet-3-bedroom-3-bathroom-4-garage-farmhouse-cottage-craftsman-country-sp289131",
  "https://www.houseplans.com/plan/2084-square-feet-3-bedroom-2-5-bathroom-3-garage-modern-country-traditional-farmhouse-sp314566",
  "https://www.houseplans.com/plan/714-square-feet-3-bedroom-2-5-bathroom-0-garage-farmhouse-modern-country-sp267839",
  "https://www.houseplans.com/plan/2417-square-feet-4-bedroom-2-bathroom-1-garage-farmhouse-southern-country-sp322721",
  "https://www.houseplans.com/plan/2782-square-feet-4-bedroom-3-bathroom-4-garage-farmhouse-modern-southern-sp325176",
  "https://www.houseplans.com/plan/3374-square-feet-3-bedroom-3-5-bathroom-0-garage-farmhouse-bungalow-39288",
  "https://www.houseplans.com/plan/2250-square-feet-3-bedroom-3-5-bathroom-4-garage-country-sp302138",
  "https://www.houseplans.com/plan/2311-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-craftsman-country-sp292271",
  "https://www.houseplans.com/plan/2720-square-feet-3-bedroom-2-5-bathroom-0-garage-farmhouse-39090",
  "https://www.houseplans.com/plan/2693-square-feet-3-bedroom-3-bathroom-4-garage-country-sp300174",
  "https://www.houseplans.com/plan/2312-square-feet-3-bedroom-2-5-bathroom-0-garage-country-sp302142",
  "https://www.houseplans.com/plan/3665-square-feet-2-bedroom-2-bathroom-6-garage-traditional-contemporary-farmhouse-sp271754",
  "https://www.houseplans.com/plan/3569-square-feet-4-bedroom-3-bathroom-3-garage-farmhouse-sp284963",
  "https://www.houseplans.com/plan/4724-square-feet-5-bedroom-4-5-bathroom-4-garage-country-colonial-sp222154",
  "https://www.houseplans.com/plan/2779-square-feet-4-bedroom-3-bathroom-4-garage-country-sp301670",
  "https://www.houseplans.com/plan/2456-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-craftsman-country-cottage-sp289132",
  "https://www.houseplans.com/plan/2937-square-feet-4-bedroom-4-5-bathroom-3-garage-country-sp301709",
  "https://www.houseplans.com/plan/3473-square-feet-3-bedroom-3-bathroom-6-garage-country-sp299137",
  "https://www.houseplans.com/plan/2350-square-feet-3-bedroom-2-5-bathroom-1-garage-cottage-sp300436",
  "https://www.houseplans.com/plan/1986-square-feet-3-bedroom-2-5-bathroom-2-garage-country-sp301708",
  "https://www.houseplans.com/plan/4601-square-feet-3-bedroom-3-5-bathroom-4-garage-country-sp302139",
  "https://www.houseplans.com/plan/3043-square-feet-3-bedroom-2-5-bathroom-4-garage-country-sp325243",
  "https://www.houseplans.com/plan/3033-square-feet-5-bedroom-3-bathroom-1-garage-farmhouse-modern-sp325884",
  "https://www.houseplans.com/plan/3414-square-feet-3-bedroom-3-5-bathroom-3-garage-farmhouse-country-sp220083",
  "https://www.houseplans.com/plan/2073-square-feet-3-bedroom-2-5-bathroom-2-garage-country-farmhouse-craftsman-sp259944",
  "https://www.houseplans.com/plan/3563-square-feet-4-bedroom-3-bathroom-3-garage-southern-country-traditional-sp294528",
  "https://www.houseplans.com/plan/2779-square-feet-3-bedroom-3-5-bathroom-4-garage-country-sp300162",
  "https://www.houseplans.com/plan/3599-square-feet-3-bedroom-3-5-bathroom-2-garage-farmhouse-38958",
  "https://www.houseplans.com/plan/2640-square-feet-1-bedroom-2-bathroom-4-garage-farmhouse-traditional-cabin-sp271637",
  "https://www.houseplans.com/plan/1410-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-country-craftsman-sp291281",
  "https://www.houseplans.com/plan/4913-square-feet-4-bedroom-3-5-bathroom-4-garage-country-sp296093",
  "https://www.houseplans.com/plan/2765-square-feet-3-bedroom-3-bathroom-3-garage-country-sp296096",
  "https://www.houseplans.com/plan/3055-square-feet-3-bedroom-4-bathroom-3-garage-country-sp299119",
  "https://www.houseplans.com/plan/3164-square-feet-3-bedroom-3-bathroom-2-garage-country-sp299121",
  "https://www.houseplans.com/plan/3004-square-feet-3-bedroom-2-5-bathroom-2-garage-country-sp301666",
  "https://www.houseplans.com/plan/2425-square-feet-3-bedroom-2-5-bathroom-2-garage-contemporary-farmhouse-sp259186",
  "https://www.houseplans.com/plan/1871-square-feet-2-bedroom-2-bathroom-3-garage-farmhouse-country-sp220307",
  "https://www.houseplans.com/plan/4261-square-feet-3-bedroom-3-5-bathroom-0-garage-farmhouse-ranch-country-sp257722",
  "https://www.houseplans.com/plan/3219-square-feet-3-bedroom-3-00-bathroom-2-garage-prairie-craftsman-farmhouse-country-sp154712",
  "https://www.houseplans.com/plan/10240-square-feet-3-bedroom-3-bathroom-18-garage-farmhouse-country-craftsman-sp271638",
  "https://www.houseplans.com/plan/3172-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-cottage-craftsman-country-sp289130",
  "https://www.houseplans.com/plan/2511-square-feet-4-bedroom-3-bathroom-4-garage-country-sp296247",
  "https://www.houseplans.com/plan/2160-square-feet-4-bedroom-2-5-bathroom-2-garage-modern-sp259872",
  "https://www.houseplans.com/plan/2887-square-feet-3-bedroom-2-5-bathroom-3-garage-country-sp301712",
  "https://www.houseplans.com/plan/2298-square-feet-2-bedroom-2-bathroom-0-garage-cabin-cottage-sp270008",
  "https://www.houseplans.com/plan/3177-square-feet-3-bedroom-2-5-bathroom-3-garage-farmhouse-country-sp291385",
  "https://www.houseplans.com/plan/2992-square-feet-4-bedroom-3-5-bathroom-0-garage-farmhouse-traditional-country-sp289787",
  "https://www.houseplans.com/plan/3177-square-feet-3-bedroom-2-5-bathroom-3-garage-farmhouse-country-sp291385",
  "https://www.houseplans.com/plan/2992-square-feet-4-bedroom-3-5-bathroom-0-garage-farmhouse-traditional-country-sp289787",
  "https://www.houseplans.com/plan/982-square-feet-4-bedroom-3-5-bathroom-0-garage-farmhouse-southern-country-craftsman-sp298799",
  "https://www.houseplans.com/plan/3150-square-feet-4-bedrooms-2-bathroom-modern-house-plan-0-garage-36019",
  "https://www.houseplans.com/plan/6720-square-feet-6-bedroom-5-bathroom-5-garage-traditional-country-farmhouse-ranch-sp291055",
  "https://www.houseplans.com/plan/3033-square-feet-3-bedroom-2-5-bathroom-2-garage-country-sp296196",
  "https://www.houseplans.com/plan/3484-square-feet-5-bedroom-3-5-bathroom-2-garage-country-sp296197",
  "https://www.houseplans.com/plan/5024-square-feet-4-bedroom-4-bathroom-5-garage-country-sp296253",
  "https://www.houseplans.com/plan/5024-square-feet-4-bedroom-4-bathroom-5-garage-country-sp296253",
  "https://www.houseplans.com/plan/3399-square-feet-3-bedroom-3-bathroom-3-garage-country-sp299120",
  "https://www.houseplans.com/plan/2400-square-feet-3-bedroom-2-5-bathroom-3-garage-country-farmhouse-sp302338",
  "https://www.houseplans.com/plan/4121-square-feet-3-bedroom-3-bathroom-4-garage-country-sp314981",
  "https://www.houseplans.com/plan/1917-square-feet-3-bedroom-2-5-bathroom-3-garage-farmhouse-country-contemporary-modern-sp154467",
  "https://www.houseplans.com/plan/2580-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-traditional-modern-sp217761",
  "https://www.houseplans.com/plan/1728-square-feet-3-bedroom-2-bathroom-0-garage-country-sp325828",
  "https://www.houseplans.com/plan/4590-square-feet-6-bedroom-4-bathroom-1-garage-farmhouse-modern-sp325886",
  "https://www.houseplans.com/plan/3162-square-feet-4-bedroom-3-bathroom-0-garage-cottage-farmhouse-craftsman-sp216033",
  "https://www.houseplans.com/plan/1500-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-13947",
  "https://www.houseplans.com/plan/2113-square-feet-4-bedroom-3-bathroom-1-garage-farmhouse-sp330884",
  "https://www.houseplans.com/plan/2717-square-feet-3-bedroom-3-5-bathroom-3-garage-farmhouse-sp299122",
  "https://www.houseplans.com/plan/3118-square-feet-2-bedroom-2-bathroom-6-garage-contemporary-modern-sp214571",
  "https://www.houseplans.com/plan/3038-square-feet-3-bedrooms-2-5-bathroom-farm-house-plans-2-garage-37601",
  "https://www.houseplans.com/plan/4072-square-feet-3-bedroom-3-5-bathroom-0-garage-country-farmhouse-traditional-sp215848",
  "https://www.houseplans.com/plan/2875-square-feet-4-bedroom-3-5-bathroom-2-garage-country-farmhouse-sp302832",
  "https://www.houseplans.com/plan/2180-square-feet-2-bedroom-2-bathroom-0-garage-ranch-country-farmhouse-sp293392",
  "https://www.houseplans.com/plan/1469-square-feet-2-bedroom-3-bathroom-0-garage-country-southern-farmhouse-sp300833",
  "https://www.houseplans.com/plan/1578-square-feet-3-bedroom-3-bathroom-0-garage-farmhouse-ranch-southern-country-sp300948",
  "https://www.houseplans.com/plan/2054-square-feet-3-bedroom-2-5-bathroom-2-garage-barndominium-traditional-sp336943",
  "https://www.houseplans.com/plan/2159-square-feet-3-bedroom-3-5-bathroom-0-garage-barndominium-farmhouse-country-southern-sp348108",
  "https://www.houseplans.com/plan/2175-square-feet-4-bedroom-3-bathroom-0-garage-barndominium-ranch-country-farmhouse-sp348160",
  "https://www.houseplans.com/plan/4528-square-feet-7-bedroom-7-bathroom-3-garage-barndominium-sp348138",
  "https://www.houseplans.com/plan/1242-square-feet-2-bedroom-2-5-bathroom-6-garage-barndominium-country-ranch-farmhouse-sp348155",
  "https://www.houseplans.com/plan/2556-square-feet-3-bedroom-2-5-bathroom-1-garage-barndominium-farmhouse-sp348031",
  "https://www.houseplans.com/plan/2228-square-feet-4-bedroom-3-bathroom-3-garage-farmhouse-barndominium-sp347737",
  "https://www.houseplans.com/plan/1400-square-feet-1-bedroom-2-bathroom-3-garage-barndominium-farmhouse-country-contemporary-sp348106",
  "https://www.houseplans.com/plan/2218-square-feet-3-bedroom-3-bathroom-2-garage-barndominium-sp345995",
  "https://www.houseplans.com/plan/2152-square-feet-4-bedroom-2-5-bathroom-3-garage-country-sp296304",
  "https://www.houseplans.com/plan/3340-square-feet-3-bedroom-2-5-bathroom-2-garage-country-modern-farmhouse-sp322168",
  "https://www.houseplans.com/plan/2010-square-feet-3-bedroom-2-bathroom-2-garage-country-sp328749",
  "https://www.houseplans.com/plan/2588-square-feet-3-bedroom-3-5-bathroom-6-garage-farmhouse-sp324650country-traditional-farmhouse-sp322167",
  "https://www.houseplans.com/plan/2466-square-feet-3-bedroom-2-5-bathroom-0-garage-farmhouse-sp325833",
  "https://www.houseplans.com/plan/2134-square-feet-3-bedroom-2-5-bathroom-0-garage-country-farmhouse-38336",
  "https://www.houseplans.com/plan/3374-square-feet-3-bedroom-3-5-bathroom-0-garage-farmhouse-bungalow-    ",
  "https://www.houseplans.com/plan/1295-square-feet-2-bedroom-2-bathroom-3-garage-farmhouse-country-ranch-southern-sp31853739288",
  "https://www.houseplans.com/plan/3068-square-feet-3-bedroom-3-5-bathroom-2-garage-ranch-farmhouse-39420",
  "https://www.houseplans.com/plan/896-square-feet-2-bedroom-1-bathroom-2-garage-country-sp321779",
  "https://www.houseplans.com/plan/2250-square-feet-3-bedroom-3-5-bathroom-4-garage-country-sp302138",
  "https://www.houseplans.com/plan/2311-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-craftsman-country-sp292271",
  "https://www.houseplans.com/plan/2637-square-feet-3-bedroom-3-bathroom-0-garage-farmhouse-cottage-country-sp302249",
  "https://www.houseplans.com/plan/2720-square-feet-3-bedroom-2-5-bathroom-0-garage-farmhouse-39090",
  "https://www.houseplans.com/plan/1596-square-feet-3-bedroom-2-bathroom-2-garage-cottage-craftsman-traditional-sp296029",
  "https://www.houseplans.com/plan/4601-square-feet-3-bedroom-3-5-bathroom-4-garage-country-sp302139",
  "https://www.houseplans.com/plan/1945-square-feet-1-bedroom-2-5-bathroom-2-garage-country-ranch-sp314521",
  "https://www.houseplans.com/plan/3043-square-feet-3-bedroom-2-5-bathroom-4-garage-country-sp325243",
  "https://www.houseplans.com/plan/3033-square-feet-5-bedroom-3-bathroom-1-garage-farmhouse-modern-sp325884",
  "https://www.houseplans.com/plan/3414-square-feet-3-bedroom-3-5-bathroom-3-garage-farmhouse-country-sp220083",
  "https://www.houseplans.com/plan/2073-square-feet-3-bedroom-2-5-bathroom-2-garage-country-farmhouse-craftsman-sp259944",
  "https://www.houseplans.com/plan/3563-square-feet-4-bedroom-3-bathroom-3-garage-southern-country-traditional-sp294528",
  "https://www.houseplans.com/plan/1631-square-feet-3-bedroom-2-bathroom-2-garage-country-ranch-sp296341",
  "https://www.houseplans.com/plan/2779-square-feet-3-bedroom-3-5-bathroom-4-garage-country-sp300162",
  "https://www.houseplans.com/plan/2456-square-feet-3-bedroom-2-5-bathroom-4-garage-country-sp325250",
  "https://www.houseplans.com/plan/2221-square-feet-3-bedroom-2-5-bathroom-4-garage-farmhouse-country-ranch-sp331763",
  "https://www.houseplans.com/plan/1916-square-feet-2-bedroom-1-bathroom-2-garage-cottage-38956",
  "https://www.houseplans.com/plan/3599-square-feet-3-bedroom-3-5-bathroom-2-garage-farmhouse-38958",
  "https://www.houseplans.com/plan/2640-square-feet-1-bedroom-2-bathroom-4-garage-farmhouse-traditional-cabin-sp271637",
  "https://www.houseplans.com/plan/1410-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-country-craftsman-sp291281",
  "https://www.houseplans.com/plan/4913-square-feet-4-bedroom-3-5-bathroom-4-garage-country-sp296093",
  "https://www.houseplans.com/plan/2765-square-feet-3-bedroom-3-bathroom-3-garage-country-sp296096",
  "https://www.houseplans.com/plan/1295-square-feet-2-bedroom-2-bathroom-3-garage-farmhouse-country-ranch-southern-sp31853739288",
  "https://www.houseplans.com/plan/3164-square-feet-3-bedroom-3-bathroom-2-garage-country-sp299121",
  "https://www.houseplans.com/plan/2048-square-feet-2-bedroom-2-5-bathroom-4-garage-country-sp300163",
  "https://www.houseplans.com/plan/3004-square-feet-3-bedroom-2-5-bathroom-2-garage-country-sp301666",
  "https://www.houseplans.com/plan/1614-square-feet-2-bedroom-2-bathroom-2-garage-country-sp303161",
  "https://www.houseplans.com/plan/4665-square-feet-5-bedroom-4-5-bathroom-2-garage-country-ranch-sp314510",
  "https://www.houseplans.com/plan/1881-square-feet-2-bedroom-2-bathroom-3-garage-country-sp321778",
  "https://www.houseplans.com/plan/2425-square-feet-3-bedroom-2-5-bathroom-2-garage-contemporary-farmhouse-sp259186",
  "https://www.houseplans.com/plan/792-square-feet-1-bedrooms-1-bathroom-cape-cod-house-plans-2-garage-32013",
  "https://www.houseplans.com/plan/1871-square-feet-2-bedroom-2-bathroom-3-garage-farmhouse-country-sp220307",
  "https://www.houseplans.com/plan/1500-square-feet-3-bedroom-2-bathroom-1-garage-farmhouse-traditional-sp294598",
  "https://www.houseplans.com/plan/4261-square-feet-3-bedroom-3-5-bathroom-0-garage-farmhouse-ranch-country-sp257722",
  "https://www.houseplans.com/plan/3219-square-feet-3-bedroom-3-00-bathroom-2-garage-prairie-craftsman-farmhouse-country-sp154712",
  "https://www.houseplans.com/plan/2062-square-feet-3-bedroom-2-5-bathroom-2-garage-barndominium-traditional-sp336945",
  "https://www.houseplans.com/plan/4117-square-feet-3-bedroom-2-bathroom-4-garage-country-sp296302",
  "https://www.houseplans.com/plan/2826-square-feet-3-bedroom-2-5-bathroom-2-garage-country-sp301711",
  "https://www.houseplans.com/plan/3063-square-feet-3-bedroom-3-bathroom-0-garage-country-sp303162",
  "https://www.houseplans.com/plan/3502-square-feet-3-bedroom-2-5-bathroom-3-garage-country-sp314980",
  "https://www.houseplans.com/plan/4013-square-feet-4-bedroom-3-5-bathroom-3-garage-barndominium-farmhouse-country-sp346363",
  "https://www.houseplans.com/plan/3869-square-feet-4-bedroom-3-5-bathroom-2-garage-farmhouse-country-craftsman-sp317786",
  "https://www.houseplans.com/plan/1973-square-feet-3-bedroom-2-5-bathroom-1-garage-farmhouse-sp325875",
  "https://www.houseplans.com/plan/3716-square-feet-3-bedroom-3-5-bathroom-0-garage-farmhouse-sp325800",
  "https://www.houseplans.com/plan/3820-square-feet-6-bedroom-4-bathroom-1-garage-farmhouse-modern-sp325887",
  "https://www.houseplans.com/plan/4073-square-feet-3-bedroom-3-bathroom-4-garage-barndominium-craftsman-farmhouse-sp345705",
  "https://www.houseplans.com/plan/1749-square-feet-2-bedroom-2-bathroom-1-garage-cottage-modern-cabin-sp291399",
  "https://www.houseplans.com/plan/2042-square-feet-3-bedroom-2-5-bathroom-3-garage-farmhouse-country-sp324529",
  "https://www.houseplans.com/plan/4624-square-feet-7-bedroom-5-bathroom-0-garage-traditional-ranch-southern-sp294517",
  "https://www.houseplans.com/plan/2084-square-feet-3-bedroom-2-5-bathroom-2-garage-barndominium-modern-farmhouse-sp333664",
  "https://www.houseplans.com/plan/2302-square-feet-4-bedroom-2-bathroom-3-garage-cabin-barndominium-farmhouse-country-sp346747",
  "https://www.houseplans.com/plan/3115-square-feet-4-bedroom-2-5-bathroom-3-garage-barndominium-farmhouse-country-sp347531",
  "https://www.houseplans.com/plan/1384-square-feet-2-bedroom-2-5-bathroom-1-garage-contemporary-sp325798",
  "https://www.houseplans.com/plan/2972-square-feet-4-bedroom-4-5-bathroom-2-garage-barndominium-farmhouse-craftsman-cabin-sp345717",
  "https://www.houseplans.com/plan/1325-square-feet-1-bedroom-2-bathroom-2-garage-farmhouse-sp325811",
  "https://www.houseplans.com/plan/3048-square-feet-4-bedroom-3-bathroom-1-garage-farmhouse-modern-sp325891",
  "https://www.houseplans.com/plan/2425-square-feet-3-bedroom-2-5-bathroom-0-garage-country-ranch-southern-farmhouse-sp331840",
  "https://www.houseplans.com/plan/1374-square-feet-1-bedroom-2-5-bathroom-3-garage-barndominium-farmhouse-traditional-sp331890",
  "https://www.houseplans.com/plan/1488-square-feet-2-bedroom-2-bathroom-3-garage-barndominium-country-craftsman-sp336903",
  "https://www.houseplans.com/plan/2682-square-feet-3-bedroom-3-5-bathroom-2-garage-country-sp314982",
  "https://www.houseplans.com/plan/3116-square-feet-3-bedroom-2-5-bathroom-2-garage-barndominium-country-farmhouse-sp346849",
  "https://www.houseplans.com/plan/1734-square-feet-1-bedroom-2-bathroom-0-garage-cabin-sp325506",
  "https://www.houseplans.com/plan/1369-square-feet-2-bedroom-2-5-bathroom-1-garage-contemporary-sp325796",
  "https://www.houseplans.com/plan/2528-square-feet-4-bedroom-3-bathroom-1-garage-farmhouse-sp325853",
  "https://www.houseplans.com/plan/2190-square-feet-3-bedroom-2-5-bathroom-0-garage-contemporary-sp325854",
  "https://www.houseplans.com/plan/2967-square-feet-3-bedroom-3-5-bathroom-1-garage-farmhouse-sp325856",
  "https://www.houseplans.com/plan/3971-square-feet-5-bedroom-4-5-bathroom-1-garage-farmhouse-sp325877",
  "https://www.houseplans.com/plan/2148-square-feet-3-bedroom-2-5-bathroom-0-garage-farmhouse-modern-sp325895",
  "https://www.houseplans.com/plan/3793-square-feet-5-bedroom-3-5-bathroom-1-garage-farmhouse-modern-sp325896",
  "https://www.houseplans.com/plan/2673-square-feet-5-bedroom-3-5-bathroom-1-garage-southern-traditional-sp325907",
  "https://www.houseplans.com/plan/4576-square-feet-3-bedroom-3-5-bathroom-1-garage-modern-sp325908",
  "https://www.houseplans.com/plan/4792-square-feet-4-bedroom-3-5-bathroom-1-garage-southern-traditional-sp325909",
  "https://www.houseplans.com/plan/2432-square-feet-2-bedroom-3-bathroom-0-garage-cottage-country-modern-farmhouse-sp216027",
  "https://www.houseplans.com/plan/2814-square-feet-2-bedroom-2-5-bathroom-2-garage-farmhouse-country-sp346105",
  "https://www.houseplans.com/plan/3023-square-feet-3-bedroom-2-5-bathroom-5-garage-farmhouse-sp324294",
  "https://www.houseplans.com/plan/2709-square-feet-3-bedroom-3-5-bathroom-7-garage-farmhouse-sp324649",
  "https://www.houseplans.com/plan/3041-square-feet-3-bedroom-2-5-bathroom-2-garage-barndominium-traditional-sp336944",
  "https://www.houseplans.com/plan/1879-square-feet-3-bedroom-2-bathroom-2-garage-bungalow-38846",
  "https://www.houseplans.com/plan/1642-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-modern-sp289231",
  "https://www.houseplans.com/plan/1362-square-feet-2-bedrooms-2-bathroom-country-house-plans-0-garage-16535",
  "https://www.houseplans.com/plan/624-square-feet-1-bedroom-1-bathroom-0-garage-cottage-craftsman-bungalow-cabin-sp268947",
  "https://www.houseplans.com/plan/624-square-feet-1-bedroom-1-bathroom-0-garage-cottage-craftsman-bungalow-cabin-sp268947",
  "https://www.houseplans.com/plan/624-square-feet-1-bedroom-1-bathroom-0-garage-cottage-craftsman-bungalow-cabin-sp268947",
  "https://www.houseplans.com/plan/2234-square-feet-3-bedroom-2-50-bathroom-2-garage-bungalow-cabin-craftsman-country-sp124126",
  "https://www.houseplans.com/plan/640-square-feet-1-bedroom-1-bathroom-2-garage-craftsman-bungalow-traditional-cottage-sp324895",
  "https://www.houseplans.com/plan/1024-square-feet-1-bedroom-1-5-bathroom-0-garage-farmhouse-beach-bungalow-cottage-sp268945",
  "https://www.houseplans.com/plan/1200-square-feet-3-bedrooms-2-bathroom-bungalow-house-plans-0-garage-36851",
  "https://www.houseplans.com/plan/1473-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-bungalow-country-ranch-sp94191",
  "https://www.houseplans.com/plan/800-square-feet-2-bedroom-1-bathroom-0-garage-cabin-cottage-sp269662",
  "https://www.houseplans.com/plan/2116-square-feet-4-bedrooms-3-bathroom-traditional-house-plans-0-garage-33516",
  "https://www.houseplans.com/plan/3742-square-feet-4-bedroom-4-00-bathroom-2-garage-prairie-craftsman-bungalow-sp154711",
  "https://www.houseplans.com/plan/2988-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-cottage-bungalow-traditional-sp289124",
  "https://www.houseplans.com/plan/1374-square-feet-3-bedrooms-2-bathroom-country-house-plans-0-garage-35856",
  "https://www.houseplans.com/plan/2128-square-feet-4-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp154443",
  "https://www.houseplans.com/plan/624-square-feet-1-bedrooms-1-bathroom-traditional-house-plans-2-garage-4474",
  "https://www.houseplans.com/plan/1440-square-feet-2-bedroom-2-bathroom-2-garage-craftsman-farmhouse-bungalow-sp289206",
  "https://www.houseplans.com/plan/1587-square-feet-3-bedroom-2-bathroom-0-garage-craftsman-southern-traditional-country-sp263841",
  "https://www.houseplans.com/plan/742-square-feet-1-bedroom-1-bathroom-0-garage-cabin-bungalow-european-modern-sp272070",
  "https://www.houseplans.com/plan/796-square-feet-1-bedroom-1-bathroom-0-garage-cottage-bungalow-contemporary-beach-sp270837",
  "https://www.houseplans.com/plan/1216-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-31657",
  "https://www.houseplans.com/plan/695-square-feet-1-bedroom-1-bathroom-0-garage-cabin-cottage-bungalow-ranch-sp270526",
  "https://www.houseplans.com/plan/1376-square-feet-3-bedrooms-2-bathroom-traditional-house-plans-0-garage-24453",
  "https://www.houseplans.com/plan/1600-square-feet-3-bedroom-2-bathroom-0-garage-bungalow-traditional-cottage-craftsman-sp256848",
  "https://www.houseplans.com/plan/1914-square-feet-3-bedroom-2-bathroom-2-garage-cottage-farmhouse-craftsman-bungalow-sp314572",
  "https://www.houseplans.com/plan/1657-square-feet-3-bedroom-2-bathroom-2-garage-bungalow-cottage-craftsman-farmhouse-sp319339",
  "https://www.houseplans.com/plan/1749-square-feet-3-bedrooms-2-bathroom-bungalow-house-plans-2-garage-37159",
  "https://www.houseplans.com/plan/1905-square-feet-3-bedrooms-2-5-bathroom-traditional-house-plans-0-garage-37324",
  "https://www.houseplans.com/plan/1260-square-feet-3-bedrooms-2-bathroom-bungalow-house-plans-0-garage-33517",
  "https://www.houseplans.com/plan/1788-square-feet-3-bedroom-2-bathroom-2-garage-farmhouse-contemporary-bungalow-sp289216",
  "https://www.houseplans.com/plan/1905-square-feet-3-bedrooms-2-5-bathroom-traditional-house-plans-0-garage-37324",
  "https://www.houseplans.com/plan/1260-square-feet-3-bedrooms-2-bathroom-bungalow-house-plans-0-garage-33517",
  "https://www.houseplans.com/plan/1428-square-feet-3-bedroom-2-bathroom-0-garage-craftsman-ranch-traditional-bungalow-sp256807",
  "https://www.houseplans.com/plan/1428-square-feet-3-bedroom-2-bathroom-0-garage-craftsman-ranch-traditional-bungalow-sp256807",
  "https://www.houseplans.com/plan/1600-square-feet-3-bedroom-2-bathroom-0-garage-bungalow-traditional-cottage-craftsman-sp256848",
  "https://www.houseplans.com/plan/2220-square-feet-4-bedroom-3-bathroom-2-garage-country-bungalow-farmhouse-craftsman-sp257330",
  "https://www.houseplans.com/plan/4343-square-feet-1-bedroom-2-bathroom-0-garage-beach-bungalow-sp273890",
  "https://www.houseplans.com/plan/1537-square-feet-2-bedrooms-2-bathroom-traditional-house-plans-2-garage-32801",
  "https://www.houseplans.com/plan/1671-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-ranch-bungalow-country-sp271580",
  "https://www.houseplans.com/plan/1874-square-feet-3-bedrooms-2-bathroom-country-house-plans-0-garage-37385",
  "https://www.houseplans.com/plan/682-square-feet-2-bedroom-1-bathroom-1-garage-modern-cabin-contemporary-sp299640",
  "https://www.houseplans.com/plan/2424-square-feet-3-bedroom-2-5-bathroom-0-garage-contemporary-bungalow-modern-sp289218",
  "https://www.houseplans.com/plan/2886-square-feet-4-bedroom-2-5-bathroom-0-garage-farmhouse-contemporary-bungalow-sp291754",
  "https://www.houseplans.com/plan/1947-square-feet-4-bedroom-2-5-bathroom-2-garage-european-contemporary-modern-bungalow-sp318381",
  "https://www.houseplans.com/plan/2004-square-feet-4-bedroom-2-5-bathroom-2-garage-farmhouse-southern-craftsman-country-sp302220",
  "https://www.houseplans.com/plan/1590-square-feet-3-bedrooms-2-bathroom-cottage-house-plans-0-garage-37095",
  "https://www.houseplans.com/plan/2092-square-feet-4-bedroom-2-5-bathroom-1-garage-craftsman-farmhouse-cottage-bungalow-sp256850",
  "https://www.houseplans.com/plan/1495-square-feet-4-bedrooms-2-bathroom-bungalow-house-plans-0-garage-10890",
  "https://www.houseplans.com/plan/1648-square-feet-4-bedroom-2-bathroom-1-garage-bungalow-traditional-sp272392",
  "https://www.houseplans.com/plan/723-square-feet-1-bedroom-1-bathroom-0-garage-beach-38278",
  "https://www.houseplans.com/plan/669-square-feet-1-bedroom-1-bathroom-2-garage-modern-bungalow-contemporary-sp257255",
  "https://www.houseplans.com/plan/999-square-feet-2-bedroom-2-bathroom-1-garage-craftsman-bungalow-39157",
  "https://www.houseplans.com/plan/788-square-feet-1-bedroom-1-bathroom-1-garage-craftsman-cottage-bungalow-40593",
  "https://www.houseplans.com/plan/2401-square-feet-1-bedroom-2-bathroom-1-garage-cabin-bungalow-39104",
  "https://www.houseplans.com/plan/261-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-39324",
  "https://www.houseplans.com/plan/1595-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-farmhouse-bungalow-country-sp257478",
  "https://www.houseplans.com/plan/8077-square-feet-4-bedroom-4-00-bathroom-4-garage-prairie-bungalow-craftsman-sp219109",
  "https://www.houseplans.com/plan/2218-square-feet-3-bedroom-2-5-bathroom-2-garage-modern-contemporary-craftsman-farmhouse-sp319344",
  "https://www.houseplans.com/plan/2234-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-37303",
  "https://www.houseplans.com/plan/2495-square-feet-3-bedroom-2-5-bathroom-3-garage-craftsman-39489",
  "https://www.houseplans.com/plan/1260-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-24446",
  "https://www.houseplans.com/plan/1195-square-feet-3-bedroom-2-00-bathroom-2-garage-victorian-country-bungalow-craftsman-sp154886",
  "https://www.houseplans.com/plan/2321-square-feet-4-bedroom-2-5-bathroom-2-garage-craftsman-sp284676",
  "https://www.houseplans.com/plan/1054-square-feet-1-bedroom-1-00-bathroom-1-garage-craftsman-traditional-bungalow-sp217248",
  "https://www.houseplans.com/plan/2632-square-feet-5-bedroom-3-5-bathroom-0-garage-craftsman-traditional-bungalow-40797",
  "https://www.houseplans.com/plan/1615-square-feet-3-bedroom-2-bathroom-0-garage-craftsman-bungalow-cottage-ranch-sp256793",
  "https://www.houseplans.com/plan/1451-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-traditional-bungalow-sp256806",
  "https://www.houseplans.com/plan/2590-square-feet-3-bedroom-3-bathroom-0-garage-craftsman-bungalow-farmhouse-traditional-sp272101",
  "https://www.houseplans.com/plan/1777-square-feet-3-bedroom-2-5-bathroom-0-garage-cottage-bungalow-39277",
  "https://www.houseplans.com/plan/1187-square-feet-3-bedrooms-2-bathroom-cottage-house-plans-1-garage-36470",
  "https://www.houseplans.com/plan/1499-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-ranch-bungalow-country-sp215056",
  "https://www.houseplans.com/plan/2296-square-feet-3-bedrooms-3-bathroom-craftsman-home-plans-0-garage-37068",
  "https://www.houseplans.com/plan/2453-square-feet-4-bedroom-3-bathroom-0-garage-country-bungalow-beach-sp94695",
  "https://www.houseplans.com/plan/912-square-feet-2-bedrooms-1-bathroom-bungalow-house-plans-0-garage-10723",
  "https://www.houseplans.com/plan/808-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-cabin-craftsman-cottage-sp271903",
  "https://www.houseplans.com/plan/1232-square-feet-3-bedroom-2-bathroom-1-garage-bungalow-craftsman-southern-cottage-sp295529",
  "https://www.houseplans.com/plan/999-square-feet-2-bedroom-2-bathroom-1-garage-craftsman-bungalow-traditional-39524",
  "https://www.houseplans.com/plan/800-square-feet-2-bedroom-2-bathroom-0-garage-craftsman-bungalow-sp123843",
  "https://www.houseplans.com/plan/5098-square-feet-4-bedroom-4-bathroom-2-garage-craftsman-bungalow-sp257035",
  "https://www.houseplans.com/plan/3218-square-feet-5-bedroom-4-bathroom-2-garage-craftsman-cottage-bungalow-sp257337",
  "https://www.houseplans.com/plan/1543-square-feet-2-bedroom-2-00-bathroom-2-garage-craftsman-prairie-european-bungalow-ranch-country-sp275903",
  "https://www.houseplans.com/plan/3108-square-feet-3-bedroom-3-5-bathroom-4-garage-bungalow-40473",
  "https://www.houseplans.com/plan/1167-square-feet-2-bedroom-2-00-bathroom-2-garage-ranch-bungalow-country-cottage-sp94302",
  "https://www.houseplans.com/plan/4549-square-feet-3-bedroom-2-bathroom-4-garage-traditional-bungalow-craftsman-sp271320",
  "https://www.houseplans.com/plan/3690-square-feet-4-bedroom-3-5-bathroom-4-garage-craftsman-bungalow-sp217386",
  "https://www.houseplans.com/plan/1202-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-31690",
  "https://www.houseplans.com/plan/1596-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-31704",
  "https://www.houseplans.com/plan/2281-square-feet-3-bedrooms-2-5-bathroom-northwest-house-plans-0-garage-36223",
  "https://www.houseplans.com/plan/798-square-feet-1-bedrooms-1-bathroom-northwest-house-plans-2-garage-36229",
  "https://www.houseplans.com/plan/1996-square-feet-3-bedrooms-2-bathroom-northwest-house-plans-0-garage-36328",
  "https://www.houseplans.com/plan/2397-square-feet-3-bedroom-3-bathroom-2-garage-craftsman-bungalow-41235",
  "https://www.houseplans.com/plan/1991-square-feet-3-bedroom-2-50-bathroom-2-garage-victorian-craftsman-country-bungalow-sp124120",
  "https://www.houseplans.com/plan/2294-square-feet-3-bedroom-2-bathroom-2-garage-prairie-bungalow-craftsman-41394",
  "https://www.houseplans.com/plan/2082-square-feet-3-bedroom-2-bathroom-2-garage-prairie-contemporary-bungalow-sp259061",
  "https://www.houseplans.com/plan/1953-square-feet-1-bedroom-1-5-bathroom-4-garage-craftsman-bungalow-traditional-sp289775",
  "https://www.houseplans.com/plan/820-square-feet-2-bedroom-1-bathroom-2-garage-craftsman-bungalow-beach-sp291466",
  "https://www.houseplans.com/plan/1403-square-feet-1-bedrooms-2-bathroom-beach-home-plans-1-garage-15806",
  "https://www.houseplans.com/plan/960-square-feet-1-bedrooms-1-bathroom-bungalow-house-plans-0-garage-20525",
  "https://www.houseplans.com/plan/1573-square-feet-3-bedroom-2-bathroom-2-garage-ranch-bungalow-craftsman-farmhouse-sp269435",
  "https://www.houseplans.com/plan/1732-square-feet-3-bedroom-2-bathroom-2-garage-contemporary-bungalow-modern-ranch-sp269625",
  "https://www.houseplans.com/plan/1497-square-feet-2-bedroom-2-bathroom-2-garage-ranch-bungalow-traditional-craftsman-sp269709",
  "https://www.houseplans.com/plan/1603-square-feet-3-bedroom-2-bathroom-2-garage-contemporary-prairie-bungalow-modern-sp270000",
  "https://www.houseplans.com/plan/1597-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-country-bungalow-ranch-sp270561",
  "https://www.houseplans.com/plan/1559-square-feet-2-bedroom-2-bathroom-3-garage-traditional-ranch-bungalow-european-sp276231",
  "https://www.houseplans.com/plan/1884-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-2-garage-36659",
  "https://www.houseplans.com/plan/2388-square-feet-4-bedroom-2-5-bathroom-2-garage-cottage-craftsman-bungalow-sp257186",
  "https://www.houseplans.com/plan/1628-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-0-garage-36869",
  "https://www.houseplans.com/plan/1643-square-feet-3-bedroom-2-00-bathroom-0-garage-country-craftsman-colonial-traditional-bungalow-sp257591",
  "https://www.houseplans.com/plan/2830-square-feet-4-bedrooms-3-5-bathroom-country-house-plans-2-garage-23484",
  "https://www.houseplans.com/plan/1256-square-feet-3-bedrooms-2-bathroom-southern-house-plans-2-garage-23336",
  "https://www.houseplans.com/plan/1379-square-feet-3-bedrooms-2-bathroom-country-house-plans-0-garage-35857",
  "https://www.houseplans.com/plan/691-square-feet-2-bedroom-1-bathroom-0-garage-country-bungalow-craftsman-41321",
  "https://www.houseplans.com/plan/3580-square-feet-5-bedroom-3-5-bathroom-4-garage-craftsman-bungalow-traditional-41327",
  "https://www.houseplans.com/plan/1400-square-feet-2-bedroom-2-00-bathroom-0-garage-craftsman-country-bungalow-sp303607",
  "https://www.houseplans.com/plan/4140-square-feet-5-bedroom-5-5-bathroom-3-garage-craftsman-bungalow-farmhouse-country-sp262744",
  "https://www.houseplans.com/plan/5889-square-feet-5-bedroom-5-5-bathroom-2-garage-traditional-craftsman-bungalow-sp291021",
  "https://www.houseplans.com/plan/1599-square-feet-3-bedrooms-2-5-bathroom-traditional-house-plans-0-garage-1009",
  "https://www.houseplans.com/plan/459-square-feet-1-bedrooms-1-bathroom-traditional-house-plans-2-garage-4514",
  "https://www.houseplans.com/plan/1809-square-feet-4-bedrooms-3-bathroom-country-house-plans-2-garage-4671",
  "https://www.houseplans.com/plan/1774-square-feet-3-bedrooms-2-5-bathroom-country-house-plans-2-garage-8866",
  "https://www.houseplans.com/plan/2690-square-feet-4-bedrooms-3-bathroom-craftsman-home-plans-3-garage-24228",
  "https://www.houseplans.com/plan/2704-square-feet-4-bedrooms-4-bathroom-craftsman-home-plans-3-garage-24325",
  "https://www.houseplans.com/plan/2308-square-feet-4-bedrooms-3-bathroom-craftsman-home-plans-2-garage-24412",
  "https://www.houseplans.com/plan/1271-square-feet-3-bedrooms-2-bathroom-traditional-house-plans-0-garage-24455",
  "https://www.houseplans.com/plan/1356-square-feet-3-bedroom-2-bathroom-2-garage-cottage-39502",
  "https://www.houseplans.com/plan/1195-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-sp124655",
  "https://www.houseplans.com/plan/2116-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-cottage-country-bungalow-sp123853",
  "https://www.houseplans.com/plan/1878-square-feet-3-bedroom-2-5-bathroom-2-garage-craftsman-bungalow-sp124477",
  "https://www.houseplans.com/plan/2232-square-feet-4-bedroom-3-bathroom-2-garage-cottage-craftsman-country-sp216630",
  "https://www.houseplans.com/plan/1898-square-feet-3-bedroom-3-bathroom-2-garage-cottage-craftsman-bungalow-sp268617",
  "https://www.houseplans.com/plan/682-square-feet-2-bedroom-1-bathroom-0-garage-cottage-sp269661",
  "https://www.houseplans.com/plan/2509-square-feet-4-bedroom-3-bathroom-2-garage-craftsman-bungalow-cottage-sp284678",
  "https://www.houseplans.com/plan/2300-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-30265",
  "https://www.houseplans.com/plan/1016-square-feet-2-bedroom-1-bathroom-0-garage-cottage-bungalow-sp123698",
  "https://www.houseplans.com/plan/533-square-feet-1-bedrooms-1-bathroom-farm-house-plans-2-garage-32423",
  "https://www.houseplans.com/plan/1625-square-feet-3-bedrooms-2-bathroom-northwest-house-plans-0-garage-2643",
  "https://www.houseplans.com/plan/1324-square-feet-2-bedrooms-1-bathroom-northwest-house-plans-0-garage-30764",
  "https://www.houseplans.com/plan/2814-square-feet-4-bedroom-2-5-bathroom-1-garage-farmhouse-ranch-bungalow-sp289226",
  "https://www.houseplans.com/plan/1874-square-feet-4-bedrooms-2-bathroom-traditional-house-plans-1-garage-27329",
  "https://www.houseplans.com/plan/2630-square-feet-3-bedroom-2-bathroom-2-garage-beach-bungalow-cottage-39213",
  "https://www.houseplans.com/plan/1620-square-feet-3-bedroom-2-bathroom-0-garage-bungalow-traditional-cottage-craftsman-39921",
  "https://www.houseplans.com/plan/1720-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-ranch-bungalow-sp124371",
  "https://www.houseplans.com/plan/1632-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-farmhouse-traditional-bungalow-sp256808",
  "https://www.houseplans.com/plan/1844-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-cottage-traditional-bungalow-sp256794",
  "https://www.houseplans.com/plan/2071-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-cottage-bungalow-farmhouse-sp94651",
  "https://www.houseplans.com/plan/1803-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-cottage-bungalow-farmhouse-sp94639",
  "https://www.houseplans.com/plan/2268-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-traditional-bungalow-40800",
  "https://www.houseplans.com/plan/2668-square-feet-5-bedroom-4-bathroom-0-garage-craftsman-traditional-bungalow-40794",
  "https://www.houseplans.com/plan/1260-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-0-garage-37029",
  "https://www.houseplans.com/plan/564-square-feet-1-bedroom-1-bathroom-2-garage-craftsman-bungalow-farmhouse-traditional-sp324896",
  "https://www.houseplans.com/plan/1450-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33514",
  "https://www.houseplans.com/plan/1791-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-36121",
  "https://www.houseplans.com/plan/1040-square-feet-3-bedrooms-2-5-bathroom-cottage-house-plans-0-garage-33554",
  "https://www.houseplans.com/plan/1029-square-feet-1-bedrooms-1-bathroom-bungalow-house-plans-3-garage-33557",
  "https://www.houseplans.com/plan/1605-square-feet-2-bedroom-2-bathroom-0-garage-bungalow-craftsman-38274",
  "https://www.houseplans.com/plan/2093-square-feet-3-bedrooms-3-5-bathroom-cottage-house-plans-2-garage-33289",
  "https://www.houseplans.com/plan/1792-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33003",
  "https://www.houseplans.com/plan/1833-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-32955",
  "https://www.houseplans.com/plan/200-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-cabin-39262",
  "https://www.houseplans.com/plan/200-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-cabin-39261",
  "https://www.houseplans.com/plan/1693-square-feet-4-bedrooms-2-5-bathroom-bungalow-house-plans-1-garage-31166",
  "https://www.houseplans.com/plan/1584-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-0-garage-33519",
  "https://www.houseplans.com/plan/1664-square-feet-3-bedroom-2-5-bathroom-1-garage-craftsman-bungalow-cottage-farmhouse-sp256845",
  "https://www.houseplans.com/plan/2723-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-cottage-farmhouse-bungalow-sp256846",
  "https://www.houseplans.com/plan/2668-square-feet-4-bedroom-4-bathroom-0-garage-farmhouse-bungalow-classical-sp295482",
  "https://www.houseplans.com/plan/2310-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-33522",
  "https://www.houseplans.com/plan/1447-square-feet-1-bedroom-1-50-bathroom-2-garage-bungalow-craftsman-sp184809",
  "https://www.houseplans.com/plan/2120-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-3-garage-13951",
  "https://www.houseplans.com/plan/2044-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-1-garage-31813",
  "https://www.houseplans.com/plan/1520-square-feet-3-bedrooms-2-bathroom-country-house-plans-2-garage-31815",
  "https://www.houseplans.com/plan/2100-square-feet-4-bedrooms-2-5-bathroom-european-house-plans-2-garage-31948",
  "https://www.houseplans.com/plan/2211-square-feet-3-bedrooms-2-5-bathroom-traditional-house-plans-3-garage-34509",
  "https://www.houseplans.com/plan/1641-square-feet-3-bedroom-2-bathroom-2-garage-bungalow-craftsman-37850",
  "https://www.houseplans.com/plan/1529-square-feet-3-bedroom-2-bathroom-2-garage-bungalow-craftsman-37979",
  "https://www.houseplans.com/plan/1986-square-feet-3-bedroom-2-5-bathroom-2-garage-bungalow-craftsman-38541",
  "https://www.houseplans.com/plan/960-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-cottage-40938",
  "https://www.houseplans.com/plan/930-square-feet-2-bedrooms-2-bathroom-bungalow-house-plans-0-garage-36064",
  "https://www.houseplans.com/plan/2001-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-sp297893",
  "https://www.houseplans.com/plan/2361-square-feet-3-bedroom-2-5-bathroom-0-garage-cottage-traditional-craftsman-bungalow-40490",
  "https://www.houseplans.com/plan/1871-square-feet-4-bedrooms-3-5-bathroom-country-house-plans-2-garage-36462",
  "https://www.houseplans.com/plan/1288-square-feet-3-bedrooms-2-bathroom-traditional-house-plans-2-garage-36471",
  "https://www.houseplans.com/plan/513-2065",
  "https://www.houseplans.com/plan/1147-square-feet-2-bedroom-2-bathroom-0-garage-cottage-37638",
  "https://www.houseplans.com/plan/1581-square-feet-3-bedroom-2-bathroom-2-garage-cottage-37660",
  "https://www.houseplans.com/plan/1786-square-feet-3-bedroom-2-bathroom-2-garage-37669",
  "https://www.houseplans.com/plan/1720-square-feet-3-bedroom-2-5-bathroom-2-garage-cottage-37670",
  "https://www.houseplans.com/plan/1397-square-feet-3-bedrooms-2-bathroom-cottage-house-plans-0-garage-36466",
  "https://www.houseplans.com/plan/1757-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-bungalow-cottage-sp266287",
  "https://www.houseplans.com/plan/1561-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-1-garage-28899",
  "https://www.houseplans.com/plan/1730-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28881",
  "https://www.houseplans.com/plan/1757-square-feet-3-bedrooms-2-bathroom-bungalow-house-plans-2-garage-37002",
  "https://www.houseplans.com/plan/1790-square-feet-3-bedroom-2-5-bathroom-2-garage-craftsman-modern-contemporary-sp124955",
  "https://www.houseplans.com/plan/1795-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-southern-traditional-sp295545",
  "https://www.houseplans.com/plan/2031-square-feet-3-bedroom-2-5-bathroom-2-garage-country-bungalow-farmhouse-craftsman-sp259937",
  "https://www.houseplans.com/plan/3068-square-feet-4-bedroom-3-5-bathroom-3-garage-european-bungalow-craftsman-sp262723",
  "https://www.houseplans.com/plan/2520-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-european-bungalow-sp262832",
  "https://www.houseplans.com/plan/2358-square-feet-3-bedroom-3-5-bathroom-0-garage-country-craftsman-bungalow-sp262839",
  "https://www.houseplans.com/plan/2073-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-bungalow-country-craftsman-sp262874",
  "https://www.houseplans.com/plan/1967-square-feet-4-bedroom-2-5-bathroom-3-garage-traditional-country-farmhouse-bungalow-sp262846",
  "https://www.houseplans.com/plan/2294-square-feet-4-bedroom-2-5-bathroom-2-garage-farmhouse-bungalow-craftsman-sp264076",
  "https://www.houseplans.com/plan/1980-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-24775",
  "https://www.houseplans.com/plan/1246-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-traditional-ranch-bungalow-sp257091",
  "https://www.houseplans.com/plan/2253-square-feet-4-bedrooms-3-bathroom-victorian-home-plans-0-garage-37069",
  "https://www.houseplans.com/plan/960-square-feet-2-bedrooms-2-bathroom-traditional-house-plans-0-garage-36724",
  "https://www.houseplans.com/plan/1191-square-feet-1-bedrooms-1-bathroom-bungalow-house-plans-2-garage-36934",
  "https://www.houseplans.com/plan/1452-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-country-cottage-sp124577",
  "https://www.houseplans.com/plan/2619-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-40442",
  "https://www.houseplans.com/plan/1522-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-0-garage-37094",
  "https://www.houseplans.com/plan/1138-square-feet-3-bedroom-2-00-bathroom-0-garage-craftsman-country-traditional-bungalow-sp257078",
  "https://www.houseplans.com/plan/1076-square-feet-3-bedroom-2-00-bathroom-0-garage-craftsman-traditional-country-bungalow-sp256956",
  "https://www.houseplans.com/plan/2482-square-feet-4-bedroom-3-5-bathroom-3-garage-craftsman-prairie-farmhouse-bungalow-sp325999",
  "https://www.houseplans.com/plan/2587-square-feet-3-bedroom-4-bathroom-3-garage-farmhouse-craftsman-bungalow-country-sp318424",
  "https://www.houseplans.com/plan/1583-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-country-bungalow-sp263911",
  "https://www.houseplans.com/plan/1307-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-victorian-bungalow-country-ranch-sp155222",
  "https://www.houseplans.com/plan/3369-square-feet-4-bedroom-3-00-bathroom-3-garage-ranch-bungalow-traditional-country-craftsman-sp123047",
  "https://www.houseplans.com/plan/2175-square-feet-3-bedroom-3-bathroom-0-garage-bungalow-country-40460",
  "https://www.houseplans.com/plan/3797-square-feet-4-bedroom-3-5-bathroom-3-garage-craftsman-bungalow-traditional-sp124530",
  "https://www.houseplans.com/plan/2247-square-feet-4-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218934",
  "https://www.houseplans.com/plan/2336-square-feet-4-bedroom-3-00-bathroom-2-garage-bungalow-country-craftsman-sp217793",
  "https://www.houseplans.com/plan/2750-square-feet-3-bedrooms-2-5-bathroom-northwest-house-plans-2-garage-29195",
  "https://www.houseplans.com/plan/1531-square-feet-3-bedrooms-2-bathroom-northwest-house-plans-2-garage-36361",
  "https://www.houseplans.com/plan/1489-square-feet-3-bedroom-2-bathroom-2-garage-cottage-bungalow-craftsman-41392",
  "https://www.houseplans.com/plan/780-square-feet-0-bedrooms-0-bathroom-craftsman-home-plans-2-garage-31017",
  "https://www.houseplans.com/plan/1403-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-ranch-bungalow-traditional-sp269876",
  "https://www.houseplans.com/plan/2646-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-3-garage-35119",
  "https://www.houseplans.com/plan/1540-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-1-garage-37357",
  "https://www.houseplans.com/plan/2314-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-bungalow-sp268591",
  "https://www.houseplans.com/plan/1836-square-feet-3-bedroom-2-bathroom-3-garage-farmhouse-modern-sp318266",
  "https://www.houseplans.com/plan/1816-square-feet-3-bedrooms-2-bathroom-cottage-house-plans-2-garage-31558",
  "https://www.houseplans.com/plan/1570-square-feet-3-bedroom-2-bathroom-2-garage-cottage-37690",
  "https://www.houseplans.com/plan/1147-square-feet-2-bedroom-2-bathroom-2-garage-cottage-37637",
  "https://www.houseplans.com/plan/1800-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-5548",
  "https://www.houseplans.com/plan/1912-square-feet-4-bedroom-2-50-bathroom-2-garage-craftsman-traditional-bungalow-sp295416",
  "https://www.houseplans.com/plan/1934-square-feet-4-bedroom-2-5-bathroom-1-garage-cottage-41019",
  "https://www.houseplans.com/plan/2124-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-3-garage-37336",
  "https://www.houseplans.com/plan/1940-square-feet-4-bedrooms-3-bathroom-northwest-house-plans-2-garage-36768",
  "https://www.houseplans.com/plan/3520-square-feet-4-bedrooms-4-5-bathroom-northwest-house-plans-2-garage-32240",
  "https://www.houseplans.com/plan/998-square-feet-2-bedroom-1-bathroom-0-garage-bungalow-cabin-contemporary-modern-sp316322",
  "https://www.houseplans.com/plan/1242-square-feet-3-bedroom-2-bathroom-1-garage-bungalow-ranch-sp271150",
  "https://www.houseplans.com/plan/1971-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-0-garage-11639",
  "https://www.houseplans.com/plan/1327-square-feet-3-bedroom-2-bathroom-2-garage-farmhouse-craftsman-bungalow-sp294604",
  "https://www.houseplans.com/plan/1921-square-feet-3-bedroom-2-5-bathroom-2-garage-craftsman-bungalow-39158",
  "https://www.houseplans.com/plan/2582-square-feet-5-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218955",
  "https://www.houseplans.com/plan/2904-square-feet-3-bedroom-2-5-bathroom-2-garage-bungalow-cottage-sp259901",
  "https://www.houseplans.com/plan/1860-square-feet-3-bedrooms-2-5-bathroom-traditional-house-plans-0-garage-33531",
  "https://www.houseplans.com/plan/2044-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-0-garage-23410",
  "https://www.houseplans.com/plan/5102-square-feet-4-bedroom-4-5-bathroom-4-garage-beach-bungalow-sp347831",
  "https://www.houseplans.com/plan/6209-square-feet-5-bedroom-6-5-bathroom-3-garage-contemporary-sp347995",
  "https://www.houseplans.com/plan/1633-square-feet-3-bedroom-2-5-bathroom-1-garage-cottage-bungalow-craftsman-country-sp348253",
  "https://www.houseplans.com/plan/2961-square-feet-3-bedroom-2-bathroom-1-garage-craftsman-cottage-bungalow-sp94680",
  "https://www.houseplans.com/plan/1710-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-bungalow-39153",
  "https://www.houseplans.com/plan/2138-square-feet-3-bedroom-2-5-bathroom-2-garage-38205",
  "https://www.houseplans.com/plan/1831-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-28768",
  "https://www.houseplans.com/plan/1728-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-32432",
  "https://www.houseplans.com/plan/1891-square-feet-3-bedroom-2-5-bathroom-3-garage-craftsman-ranch-bungalow-40795",
  "https://www.houseplans.com/plan/2265-square-feet-4-bedroom-3-5-bathroom-2-garage-bungalow-craftsman-38122",
  "https://www.houseplans.com/plan/1959-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-34478",
  "https://www.houseplans.com/plan/1492-square-feet-3-bedrooms-2-5-bathroom-cottage-house-plans-0-garage-33135",
  "https://www.houseplans.com/plan/1800-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28806",
  "https://www.houseplans.com/plan/5239-square-feet-5-bedroom-5-5-bathroom-3-garage-beach-bungalow-sp347856",
  "https://www.houseplans.com/plan/4870-square-feet-5-bedroom-5-5-bathroom-3-garage-beach-bungalow-cottage-sp347857",
  "https://www.houseplans.com/plan/4635-square-feet-4-bedroom-4-5-bathroom-3-garage-beach-contemporary-bungalow-southern-sp347993",
  "https://www.houseplans.com/plan/495-square-feet-0-bedroom-0-bathroom-2-garage-craftsman-beach-bungalow-cabin-sp348136",
  "https://www.houseplans.com/plan/1694-square-feet-3-bedroom-2-bathroom-3-garage-craftsman-cottage-bungalow-sp348107",
  "https://www.houseplans.com/plan/6212-square-feet-4-bedroom-3-5-bathroom-3-garage-bungalow-farmhouse-sp264316",
  "https://www.houseplans.com/plan/922-square-feet-2-bedroom-1-5-bathroom-0-garage-bungalow-craftsman-cottage-sp269917",
  "https://www.houseplans.com/plan/600-square-feet-1-bedroom-1-bathroom-0-garage-craftsman-cottage-bungalow-beach-sp317017",
  "https://www.houseplans.com/plan/2390-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-8088",
  "https://www.houseplans.com/plan/1908-square-feet-3-bedrooms-2-5-bathroom-country-house-plans-0-garage-31170",
  "https://www.houseplans.com/plan/689-square-feet-2-bedroom-1-bathroom-0-garage-craftsman-cottage-cabin-bungalow-sp316485",
  "https://www.houseplans.com/plan/1751-square-feet-2-bedroom-2-bathroom-2-garage-ranch-bungalow-traditional-craftsman-sp269706",
  "https://www.houseplans.com/plan/672-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-craftsman-cottage-sp269916",
  "https://www.houseplans.com/plan/1075-square-feet-1-bedroom-1-bathroom-1-garage-european-cottage-bungalow-sp271164",
  "https://www.houseplans.com/plan/600-square-feet-1-bedroom-1-bathroom-0-garage-modern-contemporary-bungalow-sp314500",
  "https://www.houseplans.com/plan/1749-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-bungalow-ranch-sp123312",
  "https://www.houseplans.com/plan/2086-square-feet-4-bedrooms-3-bathroom-spanish-style-home-plans-2-garage-16785",
  "https://www.houseplans.com/plan/2202-square-feet-5-bedrooms-4-bathroom-craftsman-home-plans-2-garage-143",
  "https://www.houseplans.com/plan/2287-square-feet-4-bedrooms-2-bathroom-craftsman-home-plans-3-garage-20726",
  "https://www.houseplans.com/plan/3850-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-bungalow-sp267960",
  "https://www.houseplans.com/plan/2776-square-feet-3-bedroom-3-bathroom-2-garage-craftsman-country-traditional-bungalow-sp275334",
  "https://www.houseplans.com/plan/3222-square-feet-2-bedroom-2-bathroom-2-garage-farmhouse-bungalow-cottage-sp291434",
  "https://www.houseplans.com/plan/1261-square-feet-3-bedroom-2-bathroom-2-garage-traditional-craftsman-bungalow-sp317172",
  "https://www.houseplans.com/plan/1489-square-feet-3-bedroom-2-bathroom-2-garage-traditional-craftsman-bungalow-colonial-sp322811",
  "https://www.houseplans.com/plan/4487-square-feet-2-bedroom-2-bathroom-4-garage-cottage-farmhouse-bungalow-sp271324",
  "https://www.houseplans.com/plan/2901-square-feet-4-bedroom-3-5-bathroom-3-garage-craftsman-bungalow-sp217380",
  "https://www.houseplans.com/plan/2159-square-feet-3-bedroom-2-00-bathroom-3-garage-craftsman-bungalow-victorian-sp304897",
  "https://www.houseplans.com/plan/1880-square-feet-3-bedrooms-2-bathroom-ranch-house-plans-2-garage-36649",
  "https://www.houseplans.com/plan/1762-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-36648",
  "https://www.houseplans.com/plan/1249-square-feet-0-bedrooms-2-bathroom-craftsman-home-plans-2-garage-36644",
  "https://www.houseplans.com/plan/1657-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-36647",
  "https://www.houseplans.com/plan/2377-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-31950",
  "https://www.houseplans.com/plan/1244-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-traditional-bungalow-ranch-sp269602",
  "https://www.houseplans.com/plan/1678-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-ranch-bungalow-farmhouse-sp269559",
  "https://www.houseplans.com/plan/1645-square-feet-3-bedroom-2-bathroom-2-garage-farmhouse-bungalow-ranch-country-sp269434",
  "https://www.houseplans.com/plan/1848-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-bungalow-traditional-sp257398",
  "https://www.houseplans.com/plan/1626-square-feet-0-bedrooms-1-bathroom-northwest-house-plans-2-garage-37267",
  "https://www.houseplans.com/plan/4667-square-feet-4-bedroom-4-5-bathroom-1-garage-craftsman-bungalow-country-sp217866",
  "https://www.houseplans.com/plan/1978-square-feet-2-bedrooms-2-bathroom-farm-house-plans-2-garage-36282",
  "https://www.houseplans.com/plan/3220-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-bungalow-traditional-farmhouse-sp289125",
  "https://www.houseplans.com/plan/1249-square-feet-2-bedroom-2-00-bathroom-2-garage-craftsman-country-bungalow-sp305085",
  "https://www.houseplans.com/plan/1746-square-feet-3-bedroom-2-00-bathroom-0-garage-ranch-bungalow-country-craftsman-sp154991",
  "https://www.houseplans.com/plan/1120-square-feet-2-bedrooms-1-5-bathroom-northwest-house-plans-0-garage-32753",
  "https://www.houseplans.com/plan/1905-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-24177",
  "https://www.houseplans.com/plan/1480-square-feet-4-bedrooms-2-bathroom-craftsman-home-plans-2-garage-35079",
  "https://www.houseplans.com/plan/2607-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-37355",
  "https://www.houseplans.com/plan/1878-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-ranch-bungalow-sp268612",
  "https://www.houseplans.com/plan/939-square-feet-0-bedrooms-0-bathroom-country-house-plans-2-garage-31618",
  "https://www.houseplans.com/plan/1185-square-feet-3-bedroom-1-00-bathroom-0-garage-traditional-craftsman-bungalow-sp217133",
  "https://www.houseplans.com/plan/1218-square-feet-3-bedroom-1-00-bathroom-0-garage-country-bungalow-craftsman-sp260000",
  "https://www.houseplans.com/plan/2072-square-feet-3-bedroom-2-00-bathroom-2-garage-country-bungalow-craftsman-sp262226",
  "https://www.houseplans.com/plan/1197-square-feet-2-bedroom-1-bathroom-2-garage-bungalow-craftsman-ranch-sp94690",
  "https://www.houseplans.com/plan/1178-square-feet-2-bedroom-1-bathroom-1-garage-craftsman-bungalow-sp264799",
  "https://www.houseplans.com/plan/2584-square-feet-2-bedroom-2-5-bathroom-1-garage-farmhouse-bungalow-craftsman-sp289217",
  "https://www.houseplans.com/plan/1400-square-feet-3-bedroom-1-bathroom-0-garage-modern-bungalow-contemporary-sp296269",
  "https://www.houseplans.com/plan/2760-square-feet-4-bedrooms-3-5-bathroom-versatile-2-garage-32956",
  "https://www.houseplans.com/plan/2160-square-feet-2-bedroom-2-bathroom-2-garage-bungalow-southern-traditional-classical-sp263994",
  "https://www.houseplans.com/plan/792-square-feet-2-bedroom-1-bathroom-0-garage-cottage-southern-country-bungalow-sp322151",
  "https://www.houseplans.com/plan/2436-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-3-garage-34487",
  "https://www.houseplans.com/plan/293-square-feet-5-bedroom-3-bathroom-0-garage-tudor-cottage-traditional-craftsman-sp324909",
  "https://www.houseplans.com/plan/2242-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-0-garage-34905",
  "https://www.houseplans.com/plan/2680-square-feet-4-bedrooms-3-bathroom-craftsman-home-plans-2-garage-37332",
  "https://www.houseplans.com/plan/2580-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-bungalow-40799",
  "https://www.houseplans.com/plan/1664-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-cottage-traditional-bungalow-sp256809",
  "https://www.houseplans.com/plan/2546-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-traditional-southern-bungalow-sp256841",
  "https://www.houseplans.com/plan/2255-square-feet-4-bedroom-3-5-bathroom-0-garage-craftsman-traditional-cottage-bungalow-sp256847",
  "https://www.houseplans.com/plan/1630-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-0-garage-33520",
  "https://www.houseplans.com/plan/2132-square-feet-3-bedroom-2-5-bathroom-2-garage-craftsman-bungalow-country-traditional-sp295487",
  "https://www.houseplans.com/plan/2566-square-feet-4-bedroom-3-bathroom-0-garage-bungalow-craftsman-traditional-cottage-sp324892",
  "https://www.houseplans.com/plan/2311-square-feet-3-bedrooms-3-bathroom-craftsman-home-plans-3-garage-31938",
  "https://www.houseplans.com/plan/2632-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-35311",
  "https://www.houseplans.com/plan/2381-square-feet-3-bedroom-2-bathroom-2-garage-traditional-sp321879",
  "https://www.houseplans.com/plan/1598-square-feet-1-bedrooms-1-5-bathroom-bungalow-house-plans-2-garage-31133",
  "https://www.houseplans.com/plan/892-square-feet-2-bedrooms-2-bathroom-traditional-house-plans-0-garage-36720",
  "https://www.houseplans.com/plan/2369-square-feet-5-bedroom-2-5-bathroom-2-garage-craftsman-modern-contemporary-41355",
  "https://www.houseplans.com/plan/2838-square-feet-3-bedroom-3-5-bathroom-2-garage-modern-beach-bungalow-sp332075",
  "https://www.houseplans.com/plan/1176-square-feet-3-bedrooms-2-bathroom-bungalow-house-plans-1-garage-37105",
  "https://www.houseplans.com/plan/1205-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-1-garage-28893",
  "https://www.houseplans.com/plan/1389-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28895",
  "https://www.houseplans.com/plan/1302-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33968",
  "https://www.houseplans.com/plan/2366-square-feet-3-bedrooms-2-5-bathroom-plantation-house-plans-2-garage-35960",
  "https://www.houseplans.com/plan/2080-square-feet-3-bedroom-2-5-bathroom-3-garage-ranch-38481",
  "https://www.houseplans.com/plan/1210-square-feet-2-bedroom-2-bathroom-1-garage-craftsman-bungalow-sp123316",
  "https://www.houseplans.com/plan/1760-square-feet-3-bedroom-2-bathroom-1-garage-craftsman-bungalow-sp123314",
  "https://www.houseplans.com/plan/2363-square-feet-3-bedroom-3-bathroom-1-garage-craftsman-bungalow-sp123066",
  "https://www.houseplans.com/plan/1851-square-feet-3-bedroom-2-50-bathroom-2-garage-prairie-bungalow-craftsman-sp289526",
  "https://www.houseplans.com/plan/1792-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-bungalow-ranch-sp290358",
  "https://www.houseplans.com/plan/1729-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-cottage-bungalow-40598",
  "https://www.houseplans.com/plan/999-square-feet-2-bedroom-2-bathroom-1-garage-craftsman-cottage-bungalow-40594",
  "https://www.houseplans.com/plan/1810-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-bungalow-39224",
  "https://www.houseplans.com/plan/1728-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-ranch-bungalow-sp337218",
  "https://www.houseplans.com/plan/600-square-feet-1-bedroom-1-bathroom-0-garage-modern-contemporary-bungalow-sp314723",
  "https://www.houseplans.com/plan/716-square-feet-1-bedroom-1-bathroom-0-garage-modern-contemporary-bungalow-sp314524",
  "https://www.houseplans.com/plan/959-square-feet-2-bedroom-2-bathroom-1-garage-craftsman-cottage-bungalow-beach-sp298034",
  "https://www.houseplans.com/plan/599-square-feet-1-bedroom-1-bathroom-2-garage-prairie-bungalow-craftsman-sp291594",
  "https://www.houseplans.com/plan/1121-square-feet-2-bedroom-2-bathroom-2-garage-prairie-bungalow-beach-sp264140",
  "https://www.houseplans.com/plan/791-square-feet-1-bedroom-2-bathroom-0-garage-traditional-bungalow-cabin-cottage-sp257805",
  "https://www.houseplans.com/plan/2458-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-bungalow-cottage-sp216551",
  "https://www.houseplans.com/plan/2382-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-cottage-bungalow-sp124761",
  "https://www.houseplans.com/plan/3934-square-feet-6-bedroom-4-5-bathroom-3-garage-country-farmhouse-bungalow-craftsman-sp259972",
  "https://www.houseplans.com/plan/1998-square-feet-3-bedroom-2-bathroom-2-garage-farmhouse-traditional-bungalow-sp262867",
  "https://www.houseplans.com/plan/4474-square-feet-5-bedroom-5-5-bathroom-3-garage-craftsman-bungalow-country-traditional-sp294506",
  "https://www.houseplans.com/plan/2026-square-feet-2-bedroom-2-bathroom-0-garage-bungalow-craftsman-contemporary-southern-40368",
  "https://www.houseplans.com/plan/1687-square-feet-2-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-sp218936",
  "https://www.houseplans.com/plan/1528-square-feet-2-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218943",
  "https://www.houseplans.com/plan/1592-square-feet-2-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-sp218976",
  "https://www.houseplans.com/plan/3641-square-feet-3-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-ranch-sp219068",
  "https://www.houseplans.com/plan/4608-square-feet-5-bedroom-4-5-bathroom-0-garage-country-bungalow-40455",
  "https://www.houseplans.com/plan/1967-square-feet-3-bedroom-3-bathroom-2-garage-cottage-bungalow-sp302828",
  "https://www.houseplans.com/plan/1792-square-feet-3-bedroom-3-00-bathroom-2-garage-craftsman-bungalow-ranch-country-sp275540",
  "https://www.houseplans.com/plan/1819-square-feet-3-bedroom-3-00-bathroom-2-garage-craftsman-bungalow-country-sp217784",
  "https://www.houseplans.com/plan/5061-square-feet-3-bedroom-3-5-bathroom-4-garage-craftsman-prairie-modern-bungalow-sp273694",
  "https://www.houseplans.com/plan/3735-square-feet-3-bedroom-3-5-bathroom-2-garage-farmhouse-modern-bungalow-sp274097",
  "https://www.houseplans.com/plan/1127-square-feet-2-bedroom-1-bathroom-2-garage-beach-contemporary-bungalow-sp315951",
  "https://www.houseplans.com/plan/1825-square-feet-3-bedroom-3-bathroom-0-garage-craftsman-bungalow-country-sp154669",
  "https://www.houseplans.com/plan/1859-square-feet-3-bedroom-2-5-bathroom-2-garage-bungalow-craftsman-cottage-40869",
  "https://www.houseplans.com/plan/2506-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-bungalow-sp218865",
  "https://www.houseplans.com/plan/2361-square-feet-3-bedroom-2-5-bathroom-0-garage-bungalow-traditional-cottage-40491",
  "https://www.houseplans.com/plan/2052-square-feet-3-bedroom-2-5-bathroom-2-garage-country-cottage-bungalow-craftsman-sp270249",
  "https://www.houseplans.com/plan/2125-square-feet-3-bedrooms-3-bathroom-northwest-house-plans-2-garage-31176",
  "https://www.houseplans.com/plan/2477-square-feet-3-bedrooms-2-5-bathroom-traditional-house-plans-2-garage-23396",
  "https://www.houseplans.com/plan/150-square-feet-3-bedroom-2-bathroom-0-garage-modern-bungalow-sp325183",
  "https://www.houseplans.com/plan/3688-square-feet-4-bedroom-2-bathroom-3-garage-craftsman-bungalow-sp215131",
  "https://www.houseplans.com/plan/1758-square-feet-4-bedroom-3-bathroom-2-garage-bungalow-ranch-craftsman-southern-sp322078",
  "https://www.houseplans.com/plan/1316-square-feet-2-bedroom-2-bathroom-2-garage-ranch-bungalow-cottage-country-sp323678",
  "https://www.houseplans.com/plan/1488-square-feet-3-bedrooms-2-bathroom-european-house-plans-2-garage-36645",
  "https://www.houseplans.com/plan/1798-square-feet-3-bedroom-2-bathroom-2-garage-traditional-craftsman-bungalow-39695",
  "https://www.houseplans.com/plan/1787-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28961",
  "https://www.houseplans.com/plan/1730-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28618",
  "https://www.houseplans.com/plan/1958-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-37318",
  "https://www.houseplans.com/plan/2502-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-31810",
  "https://www.houseplans.com/plan/2308-square-feet-4-bedrooms-4-bathroom-craftsman-home-plans-0-garage-33737",
  "https://www.houseplans.com/plan/2460-square-feet-3-bedrooms-3-bathroom-bungalow-house-plans-2-garage-37304",
  "https://www.houseplans.com/plan/725-square-feet-2-bedroom-1-bathroom-0-garage-cottage-craftsman-cabin-bungalow-sp325289",
  "https://www.houseplans.com/plan/6005-square-feet-4-bedroom-4-5-bathroom-3-garage-beach-bungalow-modern-sp335933",
  "https://www.houseplans.com/plan/1590-square-feet-1-bedroom-1-bathroom-3-garage-traditional-colonial-bungalow-farmhouse-sp291954",
  "https://www.houseplans.com/plan/1875-square-feet-2-bedroom-2-bathroom-2-garage-craftsman-farmhouse-bungalow-sp303150",
  "https://www.houseplans.com/plan/1552-square-feet-3-bedroom-2-bathroom-2-garage-traditional-farmhouse-bungalow-sp319322",
  "https://www.houseplans.com/plan/1495-square-feet-3-bedroom-2-bathroom-2-garage-traditional-bungalow-beach-sp319368",
  "https://www.houseplans.com/plan/1662-square-feet-3-bedroom-2-bathroom-2-garage-traditional-colonial-bungalow-sp321739",
  "https://www.houseplans.com/plan/2049-square-feet-3-bedroom-2-bathroom-2-garage-traditional-bungalow-craftsman-sp322022",
  "https://www.houseplans.com/plan/1660-square-feet-3-bedroom-2-bathroom-2-garage-traditional-cottage-bungalow-craftsman-sp322177",
  "https://www.houseplans.com/plan/1598-square-feet-3-bedroom-2-bathroom-2-garage-traditional-cottage-bungalow-craftsman-sp322184",
  "https://www.houseplans.com/plan/1489-square-feet-3-bedroom-2-bathroom-2-garage-traditional-craftsman-bungalow-colonial-sp322810",
  "https://www.houseplans.com/plan/1332-square-feet-1-bedroom-1-5-bathroom-0-garage-traditional-beach-craftsman-bungalow-sp324018",
  "https://www.houseplans.com/plan/1521-square-feet-3-bedroom-2-bathroom-2-garage-traditional-bungalow-craftsman-sp325893",
  "https://www.houseplans.com/plan/1066-square-feet-0-bedroom-0-bathroom-2-garage-traditional-craftsman-bungalow-sp331891",
  "https://www.houseplans.com/plan/3533-square-feet-4-bedroom-3-bathroom-3-garage-craftsman-bungalow-cottage-sp271166",
  "https://www.houseplans.com/plan/4993-square-feet-3-bedroom-2-5-bathroom-3-garage-craftsman-farmhouse-bungalow-sp271325",
  "https://www.houseplans.com/plan/2764-square-feet-4-bedroom-3-5-bathroom-2-garage-southern-craftsman-bungalow-sp337225",
  "https://www.houseplans.com/plan/396-square-feet-0-bedroom-1-bathroom-0-garage-craftsman-beach-bungalow-sp346225",
  "https://www.houseplans.com/plan/1734-square-feet-1-bedroom-2-bathroom-0-garage-cabin-sp325506",
  "https://www.houseplans.com/plan/1275-square-feet-2-bedroom-2-bathroom-2-garage-european-adobe-southwestern-bungalow-cottage-sp347599",
  "https://www.houseplans.com/plan/2512-square-feet-3-bedroom-3-bathroom-2-garage-beach-mediterranean-craftsman-bungalow-sp291004",
  "https://www.houseplans.com/plan/3861-square-feet-3-bedrooms-3-bathroom-northwest-house-plans-0-garage-36194",
  "https://www.houseplans.com/plan/4684-square-feet-5-bedrooms-3-5-bathroom-northwest-house-plans-0-garage-36189",
  "https://www.houseplans.com/plan/1453-square-feet-1-bedrooms-1-5-bathroom-northwest-house-plans-1-garage-36188",
  "https://www.houseplans.com/plan/3695-square-feet-3-bedrooms-3-5-bathroom-northwest-house-plans-0-garage-36186",
  "https://www.houseplans.com/plan/4120-square-feet-4-bedrooms-3-bathroom-northwest-house-plans-2-garage-36185",
  "https://www.houseplans.com/plan/3471-square-feet-3-bedrooms-4-bathroom-northwest-house-plans-2-garage-36184",
  "https://www.houseplans.com/plan/2732-square-feet-2-bedrooms-2-bathroom-traditional-house-plans-2-garage-36276",
  "https://www.houseplans.com/plan/3522-square-feet-3-bedrooms-2-5-bathroom-northwest-house-plans-2-garage-36278",
  "https://www.houseplans.com/plan/3678-square-feet-5-bedrooms-4-bathroom-country-house-plans-2-garage-36288",
  "https://www.houseplans.com/plan/4285-square-feet-3-bedrooms-3-5-bathroom-northwest-house-plans-2-garage-36289",
  "https://www.houseplans.com/plan/1600-square-feet-3-bedrooms-2-5-bathroom-cottage-house-plans-2-garage-16188",
  "https://www.houseplans.com/plan/1540-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-2-garage-16107",
  "https://www.houseplans.com/plan/1360-square-feet-2-bedroom-2-bathroom-0-garage-bungalow-cabin-cottage-sp324095",
  "https://www.houseplans.com/plan/2489-square-feet-3-bedroom-2-5-bathroom-2-garage-prairie-modern-bungalow-sp322708",
  "https://www.houseplans.com/plan/2426-square-feet-3-bedroom-2-5-bathroom-2-garage-country-bungalow-craftsman-sp284779",
  "https://www.houseplans.com/plan/2221-square-feet-3-bedroom-2-bathroom-2-garage-prairie-traditional-bungalow-sp222282",
  "https://www.houseplans.com/plan/3389-square-feet-4-bedrooms-2-5-bathroom-bungalow-house-plans-3-garage-17501",
  "https://www.houseplans.com/plan/2146-square-feet-3-bedrooms-2-5-bathroom-country-house-plans-2-garage-3606",
  "https://www.houseplans.com/plan/3103-square-feet-5-bedroom-3-50-bathroom-2-garage-craftsman-bungalow-sp304852",
  "https://www.houseplans.com/plan/1568-square-feet-3-bedrooms-2-5-bathroom-country-house-plans-2-garage-4660",
  "https://www.houseplans.com/plan/1923-square-feet-3-bedrooms-3-bathroom-bungalow-house-plans-2-garage-4675",
  "https://www.houseplans.com/plan/1376-square-feet-3-bedrooms-2-bathroom-traditional-house-plans-2-garage-24454",
  "https://www.houseplans.com/plan/2168-square-feet-3-bedrooms-3-bathroom-craftsman-home-plans-2-garage-24341",
  "https://www.houseplans.com/plan/2429-square-feet-4-bedrooms-3-bathroom-craftsman-home-plans-2-garage-24336",
  "https://www.houseplans.com/plan/2255-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-35091",
  "https://www.houseplans.com/plan/4090-square-feet-3-bedrooms-3-bathroom-craftsman-home-plans-2-garage-35171",
  "https://www.houseplans.com/plan/2045-square-feet-3-bedroom-2-5-bathroom-2-garage-modern-country-sp324273",
  "https://www.houseplans.com/plan/2630-square-feet-4-bedroom-2-bathroom-0-garage-modern-contemporary-bungalow-sp296275",
  "https://www.houseplans.com/plan/2770-square-feet-4-bedroom-2-bathroom-0-garage-farmhouse-modern-country-bungalow-sp314748",
  "https://www.houseplans.com/plan/1167-square-feet-2-bedroom-1-bathroom-0-garage-bungalow-contemporary-modern-ranch-sp315975",
  "https://www.houseplans.com/plan/1771-square-feet-3-bedroom-1-bathroom-1-garage-bungalow-sp318408",
  "https://www.houseplans.com/plan/1027-square-feet-2-bedroom-1-bathroom-1-garage-bungalow-sp318407",
  "https://www.houseplans.com/plan/1487-square-feet-2-bedroom-1-bathroom-1-garage-bungalow-traditional-contemporary-modern-sp319378",
  "https://www.houseplans.com/plan/2248-square-feet-3-bedroom-2-bathroom-0-garage-bungalow-country-modern-farmhouse-sp319366",
  "https://www.houseplans.com/plan/1487-square-feet-2-bedroom-1-bathroom-0-garage-bungalow-cabin-country-contemporary-sp316324",
  "https://www.houseplans.com/plan/2400-square-feet-3-bedroom-2-bathroom-0-garage-bungalow-farmhouse-modern-ranch-sp315976",
  "https://www.houseplans.com/plan/1803-square-feet-2-bedrooms-2-bathroom-ranch-house-plans-1-garage-12561",
  "https://www.houseplans.com/plan/2130-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-0-garage-31302",
  "https://www.houseplans.com/plan/1948-square-feet-3-bedroom-2-bathroom-0-garage-cottage-bungalow-traditional-craftsman-39661",
  "https://www.houseplans.com/plan/2715-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-4-garage-34534",
  "https://www.houseplans.com/plan/3718-square-feet-4-bedrooms-3-5-bathroom-craftsman-home-plans-3-garage-34532",
  "https://www.houseplans.com/plan/2633-square-feet-3-bedrooms-2-5-bathroom-northwest-house-plans-2-garage-36737",
  "https://www.houseplans.com/plan/1700-square-feet-3-bedrooms-2-bathroom-cottage-house-plans-2-garage-32583",
  "https://www.houseplans.com/plan/1760-square-feet-3-bedrooms-2-5-bathroom-bungalow-house-plans-2-garage-32606",
  "https://www.houseplans.com/plan/1825-square-feet-3-bedrooms-2-5-bathroom-cottage-house-plans-2-garage-34924",
  "https://www.houseplans.com/plan/1915-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-32954",
  "https://www.houseplans.com/plan/1887-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-2-garage-33002",
  "https://www.houseplans.com/plan/2297-square-feet-3-bedrooms-2-5-bathroom-craftsman-home-plans-3-garage-33069",
  "https://www.houseplans.com/plan/1587-square-feet-3-bedroom-0-bathroom-2-garage-craftsman-bungalow-traditional-sp263844",
  "https://www.houseplans.com/plan/2693-square-feet-3-bedrooms-4-5-bathroom-modern-house-plan-2-garage-36582",
  "https://www.houseplans.com/plan/4272-square-feet-4-bedroom-3-5-bathroom-2-garage-traditional-farmhouse-country-bungalow-39441",
  "https://www.houseplans.com/plan/2519-square-feet-4-bedroom-3-00-bathroom-2-garage-craftsman-bungalow-country-sp218877",
  "https://www.houseplans.com/plan/2393-square-feet-3-bedroom-3-bathroom-2-garage-craftsman-cottage-bungalow-sp347667",
  "https://www.houseplans.com/plan/2268-square-feet-4-bedroom-3-bathroom-0-garage-craftsman-cottage-bungalow-traditional-sp295483",
  "https://www.houseplans.com/plan/1750-square-feet-3-bedroom-2-5-bathroom-2-garage-craftsman-bungalow-cottage-traditional-sp295492",
  "https://www.houseplans.com/plan/3163-square-feet-4-bedroom-3-5-bathroom-0-garage-farmhouse-craftsman-traditional-sp295904",
  "https://www.houseplans.com/plan/2236-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-37366",
  "https://www.houseplans.com/plan/2737-square-feet-4-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-country-sp295309",
  "https://www.houseplans.com/plan/2468-square-feet-3-bedroom-2-50-bathroom-2-garage-bungalow-craftsman-country-sp295314",
  "https://www.houseplans.com/plan/2921-square-feet-4-bedrooms-2-5-bathroom-craftsman-home-plans-3-garage-31151",
  "https://www.houseplans.com/plan/1665-square-feet-2-bedrooms-1-5-bathroom-craftsman-home-plans-2-garage-31136",
  "https://www.houseplans.com/plan/2945-square-feet-3-bedrooms-2-5-bathroom-northwest-house-plans-3-garage-27867",
  "https://www.houseplans.com/plan/2993-square-feet-4-bedrooms-2-5-bathroom-european-house-plans-3-garage-27866",
  "https://www.houseplans.com/plan/1794-square-feet-3-bedroom-2-bathroom-0-garage-craftsman-ranch-bungalow-prairie-sp315838",
  "https://www.houseplans.com/plan/3202-square-feet-3-bedrooms-3-bathroom-craftsman-home-plans-0-garage-31156",
  "https://www.houseplans.com/plan/3286-square-feet-4-bedrooms-3-5-bathroom-traditional-house-plans-3-garage-6519",
  "https://www.houseplans.com/plan/2655-square-feet-4-bedrooms-3-bathroom-craftsman-home-plans-2-garage-33808",
  "https://www.houseplans.com/plan/1788-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28960",
  "https://www.houseplans.com/plan/1556-square-feet-3-bedrooms-2-bathroom-cottage-house-plans-2-garage-28898",
  "https://www.houseplans.com/plan/1205-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-1-garage-28892",
  "https://www.houseplans.com/plan/1205-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-1-garage-28891",
  "https://www.houseplans.com/plan/1799-square-feet-3-bedrooms-2-bathroom-country-house-plans-2-garage-28886",
  "https://www.houseplans.com/plan/1806-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-28885",
  "https://www.houseplans.com/plan/1336-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33925",
  "https://www.houseplans.com/plan/1336-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33926",
  "https://www.houseplans.com/plan/1817-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33938",
  "https://www.houseplans.com/plan/1817-square-feet-3-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33939",
  "https://www.houseplans.com/plan/1337-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33943",
  "https://www.houseplans.com/plan/1367-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33944",
  "https://www.houseplans.com/plan/1367-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33945",
  "https://www.houseplans.com/plan/1302-square-feet-2-bedrooms-2-bathroom-craftsman-home-plans-2-garage-33969",
  "https://www.houseplans.com/plan/2713-square-feet-4-bedrooms-3-bathroom-victorian-home-plans-2-garage-37062",
  "https://www.houseplans.com/plan/3825-square-feet-4-bedroom-3-5-bathroom-2-garage-craftsman-bungalow-country-beach-sp155133",
  "https://www.houseplans.com/plan/4418-square-feet-4-bedroom-4-bathroom-0-garage-traditional-craftsman-bungalow-sp216092",
  "https://www.houseplans.com/plan/2565-square-feet-4-bedroom-3-5-bathroom-2-garage-country-bungalow-craftsman-farmhouse-sp264044",
  "https://www.houseplans.com/plan/1846-square-feet-3-bedroom-2-bathroom-2-garage-farmhouse-sp314552",
  "https://www.houseplans.com/plan/2828-square-feet-4-bedroom-3-bathroom-3-garage-traditional-sp301830",
  "https://www.houseplans.com/plan/2152-square-feet-3-bedroom-2-5-bathroom-2-garage-traditional-sp301713",
  "https://www.houseplans.com/plan/1997-square-feet-3-bedrooms-2-5-bathroom-cottage-house-plans-0-garage-9428",
  "https://www.houseplans.com/plan/1602-square-feet-2-bedroom-2-bathroom-2-garage-bungalow-craftsman-38204",
  "https://www.houseplans.com/plan/684-square-feet-1-bedroom-1-bathroom-0-garage-modern-contemporary-bungalow-sp303149",
  "https://www.houseplans.com/plan/2378-square-feet-3-bedroom-2-bathroom-3-garage-craftsman-bungalow-ranch-sp337222",
  "https://www.houseplans.com/plan/1863-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-bungalow-39154",
  "https://www.houseplans.com/plan/2068-square-feet-4-bedroom-2-5-bathroom-2-garage-prairie-bungalow-craftsman-39162",
  "https://www.houseplans.com/plan/1506-square-feet-3-bedroom-2-bathroom-2-garage-traditional-bungalow-craftsman-39505",
  "https://www.houseplans.com/plan/2272-square-feet-4-bedroom-3-bathroom-2-garage-craftsman-bungalow-traditional-39522",
  "https://www.houseplans.com/plan/1693-square-feet-3-bedroom-2-5-bathroom-2-garage-bungalow-craftsman-38305",
  "https://www.houseplans.com/plan/1760-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-ranch-sp290354",
  "https://www.houseplans.com/plan/1915-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-ranch-sp290357",
  "https://www.houseplans.com/plan/2760-square-feet-4-bedroom-3-50-bathroom-2-garage-craftsman-traditional-bungalow-country-sp290360",
  "https://www.houseplans.com/plan/1534-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-bungalow-ranch-sp290364",
  "https://www.houseplans.com/plan/1825-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp290365",
  "https://www.houseplans.com/plan/1830-square-feet-3-bedroom-2-00-bathroom-2-garage-craftsman-bungalow-sp290372",
  "https://www.houseplans.com/plan/3593-square-feet-3-bedroom-3-bathroom-1-garage-craftsman-bungalow-traditional-sp123035",
  "https://www.houseplans.com/plan/2070-square-feet-3-bedroom-2-5-bathroom-2-garage-bungalow-craftsman-38337",
  "https://www.houseplans.com/plan/1589-square-feet-3-bedroom-2-50-bathroom-2-garage-country-craftsman-bungalow-sp327908",
  "https://www.houseplans.com/plan/3408-square-feet-3-bedroom-3-bathroom-2-garage-craftsman-bungalow-country-ranch-sp319311",
  "https://www.houseplans.com/plan/1846-square-feet-2-bedroom-2-5-bathroom-0-garage-cottage-bungalow-craftsman-sp318120",
  "https://www.houseplans.com/plan/5300-square-feet-6-bedroom-4-5-bathroom-2-garage-craftsman-traditional-40532",
  "https://www.houseplans.com/plan/2115-square-feet-3-bedroom-2-bathroom-1-garage-craftsman-bungalow-sp123866",
  "https://www.houseplans.com/plan/2377-square-feet-3-bedroom-3-bathroom-1-garage-traditional-bungalow-sp123209",
  "https://www.houseplans.com/plan/1704-square-feet-3-bedroom-2-00-bathroom-2-garage-bungalow-craftsman-country-sp327960",
  "https://www.houseplans.com/plan/3001-square-feet-4-bedroom-4-00-bathroom-2-garage-craftsman-bungalow-country-sp328523",
  "https://www.houseplans.com/plan/1384-square-feet-2-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218928",
  "https://www.houseplans.com/plan/2127-square-feet-3-bedroom-3-00-bathroom-2-garage-craftsman-bungalow-country-sp218929",
  "https://www.houseplans.com/plan/2484-square-feet-3-bedroom-3-00-bathroom-2-garage-craftsman-bungalow-sp218931",
  "https://www.houseplans.com/plan/2519-square-feet-3-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-sp218932",
  "https://www.houseplans.com/plan/2582-square-feet-4-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218933",
  "https://www.houseplans.com/plan/2381-square-feet-4-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218935",
  "https://www.houseplans.com/plan/2582-square-feet-4-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218937",
  "https://www.houseplans.com/plan/2977-square-feet-3-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-ranch-sp218940",
  "https://www.houseplans.com/plan/2029-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218941",
  "https://www.houseplans.com/plan/2448-square-feet-2-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-ranch-sp219007",
  "https://www.houseplans.com/plan/1948-square-feet-2-bedroom-2-00-bathroom-2-garage-bungalow-country-craftsman-sp219006",
  "https://www.houseplans.com/plan/1943-square-feet-3-bedroom-2-00-bathroom-0-garage-bungalow-country-craftsman-sp219002",
  "https://www.houseplans.com/plan/2243-square-feet-2-bedroom-2-50-bathroom-3-garage-bungalow-country-craftsman-sp218981",
  "https://www.houseplans.com/plan/3605-square-feet-3-bedroom-4-00-bathroom-2-garage-craftsman-bungalow-sp123841",
  "https://www.houseplans.com/plan/1633-square-feet-2-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-sp218970",
  "https://www.houseplans.com/plan/1724-square-feet-2-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-sp218963",
  "https://www.houseplans.com/plan/3130-square-feet-4-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218996",
  "https://www.houseplans.com/plan/1568-square-feet-2-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-sp218962",
  "https://www.houseplans.com/plan/3123-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218960",
  "https://www.houseplans.com/plan/2606-square-feet-4-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218958garage-craftsman-bungalow-sp218961",
  "https://www.houseplans.com/plan/2660-square-feet-5-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218957",
  "https://www.houseplans.com/plan/1588-square-feet-5-bedroom-3-50-bathroom-2-garage-craftsman-bungalow-sp218954garage-craftsman-bungalow-sp218956garage-",
  "https://www.houseplans.com/plan/2121-square-feet-4-bedroom-3-00-bathroom-2-garage-craftsman-bungalow-sp218949garage-craftsman-bungalow-sp218953craftsman-bungalow-sp218959",
  "https://www.houseplans.com/plan/2573-square-feet-4-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218947",
  "https://www.houseplans.com/plan/2959-square-feet-4-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218950",
  "https://www.houseplans.com/plan/1376-square-feet-2-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218945garage-craftsman-bungalow-sp218946",
  "https://www.houseplans.com/plan/2773-square-feet-5-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp218944",
  "https://www.houseplans.com/plan/2863-square-feet-4-bedroom-3-50-bathroom-2-garage-craftsman-bungalow-sp218951",
  "https://www.houseplans.com/plan/2917-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp218942",
  "https://www.houseplans.com/plan/3407-square-feet-3-bedroom-2-50-bathroom-2-garage-prairie-bungalow-craftsman-sp219037",
  "https://www.houseplans.com/plan/3434-square-feet-3-bedroom-3-50-bathroom-3-garage-craftsman-bungalow-sp219134",
  "https://www.houseplans.com/plan/3617-square-feet-3-bedroom-3-50-bathroom-2-garage-craftsman-bungalow-country-sp219070",
  "https://www.houseplans.com/plan/3001-square-feet-2-bedroom-3-50-bathroom-2-garage-craftsman-bungalow-sp219124",
  "https://www.houseplans.com/plan/3536-square-feet-2-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp219130",
  "https://www.houseplans.com/plan/3210-square-feet-3-bedroom-3-50-bathroom-2-garage-craftsman-bungalow-ranch-sp219123",
  "https://www.houseplans.com/plan/2976-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-bungalow-sp219131",
  "https://www.houseplans.com/plan/3679-square-feet-2-bedroom-2-50-bathroom-3-garage-craftsman-bungalow-traditional-sp219082",
  "https://www.houseplans.com/plan/1737-square-feet-3-bedroom-2-00-bathroom-0-garage-craftsman-bungalow-sp219133",
  "https://www.houseplans.com/plan/1259-square-feet-3-bedroom-2-50-bathroom-0-garage-craftsman-bungalow-traditional-country-sp267825",
  "https://www.houseplans.com/plan/1512-square-feet-3-bedroom-2-50-bathroom-0-garage-craftsman-country-traditional-bungalow-sp216775",
  "https://www.houseplans.com/plan/940-square-feet-2-bedroom-1-bathroom-0-garage-ranch-bungalow-sp325059",
  "https://www.houseplans.com/plan/280-square-feet-0-bedroom-1-bathroom-0-garage-bungalow-beach-cottage-sp298600",
  "https://www.houseplans.com/plan/2953-square-feet-4-bedroom-3-50-bathroom-3-garage-country-craftsman-bungalow-sp275613",
  "https://www.houseplans.com/plan/1768-square-feet-3-bedroom-2-5-bathroom-2-garage-bungalow-40603",
  "https://www.houseplans.com/plan/2336-square-feet-2-bedroom-2-bathroom-2-garage-modern-ranch-contemporary-sp296082",
  "https://www.houseplans.com/plan/2274-square-feet-4-bedroom-2-50-bathroom-2-garage-contemporary-bungalow-craftsman-sp257112",
  "https://www.houseplans.com/plan/3982-square-feet-4-bedrooms-3-5-bathroom-luxury-home-plans-3-garage-93",
  "https://www.houseplans.com/plan/4701-square-feet-4-bedrooms-3-5-bathroom-traditional-house-plans-3-garage-100",
  "https://www.houseplans.com/plan/2080-square-feet-3-bedrooms-2-bathroom-northwest-house-plans-2-garage-16780",
  "https://www.houseplans.com/plan/2845-square-feet-3-bedroom-3-5-bathroom-3-garage-farmhouse-european-craftsman-bungalow-sp345993",
  "https://www.houseplans.com/plan/1170-square-feet-3-bedroom-1-bathroom-0-garage-bungalow-sp318550",
  "https://www.houseplans.com/plan/3757-square-feet-4-bedroom-3-bathroom-2-garage-country-cottage-bungalow-farmhouse-sp273043",
  "https://www.houseplans.com/plan/1230-square-feet-3-bedrooms-2-bathroom-contemporary-house-plans-0-garage-5321",
  "https://www.houseplans.com/plan/1679-square-feet-2-bedroom-2-bathroom-0-garage-modern-contemporary-sp321732",
  "https://www.houseplans.com/plan/1200-square-feet-2-bedroom-2-bathroom-0-garage-cabin-modern-contemporary-country-sp261046",
  "https://www.houseplans.com/plan/900-square-feet-2-bedrooms-1-bathroom-country-house-plans-0-garage-23942",
  "https://www.houseplans.com/plan/1665-square-feet-2-bedroom-2-bathroom-2-garage-cabin-39488",
  "https://www.houseplans.com/plan/768-square-feet-1-bedrooms-1-bathroom-bungalow-house-plans-0-garage-15686",
  "https://www.houseplans.com/plan/1362-square-feet-2-bedrooms-2-bathroom-country-house-plans-0-garage-16535",
  "https://www.houseplans.com/plan/624-square-feet-1-bedroom-1-bathroom-0-garage-cottage-craftsman-bungalow-cabin-sp268947",
  "https://www.houseplans.com/plan/1311-square-feet-3-bedroom-2-bathroom-2-garage-ranch-country-craftsman-cabin-sp264203",
  "https://www.houseplans.com/plan/384-square-feet-1-bedroom-1-bathroom-0-garage-country-40017",
  "https://www.houseplans.com/plan/2234-square-feet-3-bedroom-2-50-bathroom-2-garage-bungalow-cabin-craftsman-country-sp124126",
  "https://www.houseplans.com/plan/600-square-feet-1-bedrooms-1-bathroom-cottage-house-plans-0-garage-25908",
  "https://www.houseplans.com/plan/1479-square-feet-3-bedrooms-2-bathroom-northwest-house-plans-0-garage-23501",
  "https://www.houseplans.com/plan/1400-square-feet-2-bedrooms-2-bathroom-cottage-house-plans-0-garage-1174",
  "https://www.houseplans.com/plan/1816-square-feet-3-bedroom-2-50-bathroom-2-garage-craftsman-country-cabin-sp217574",
  "https://www.houseplans.com/plan/2180-square-feet-3-bedroom-2-5-bathroom-0-garage-contemporary-modern-ranch-cabin-39693",
  "https://www.houseplans.com/plan/1188-square-feet-2-bedroom-2-bathroom-1-garage-modern-contemporary-cabin-cottage-sp264246",
  "https://www.houseplans.com/plan/1162-square-feet-2-bedroom-2-bathroom-0-garage-modern-european-contemporary-cabin-sp292670",
  "https://www.houseplans.com/plan/1249-square-feet-3-bedroom-2-bathroom-0-garage-cabin-beach-cottage-country-sp269656",
  "https://www.houseplans.com/plan/800-square-feet-2-bedroom-1-bathroom-0-garage-cabin-cottage-sp269662",
  "https://www.houseplans.com/plan/1655-square-feet-2-bedroom-2-5-bathroom-0-garage-modern-sp325812",
  "https://www.houseplans.com/plan/652-square-feet-2-bedroom-1-5-bathroom-1-garage-cottage-cabin-craftsman-traditional-sp267858",
  "https://www.houseplans.com/plan/1011-square-feet-2-bedroom-2-bathroom-0-garage-country-beach-cabin-cottage-sp275325",
  "https://www.houseplans.com/plan/1484-square-feet-3-bedroom-2-bathroom-1-garage-cabin-sp318541",
  "https://www.houseplans.com/plan/2055-square-feet-4-bedroom-3-bathroom-1-garage-cottage-cabin-craftsman-country-sp264232",
  "https://www.houseplans.com/plan/1230-square-feet-3-bedroom-2-00-bathroom-0-garage-sp124340",
  "https://www.houseplans.com/plan/640-square-feet-1-bedroom-1-bathroom-0-garage-modern-38327",
  "https://www.houseplans.com/plan/850-square-feet-2-bedroom-1-bathroom-0-garage-modern-cabin-40508",
  "https://www.houseplans.com/plan/1966-square-feet-4-bedroom-3-bathroom-0-garage-country-cottage-farmhouse-cabin-sp215998",
  "https://www.houseplans.com/plan/1725-square-feet-3-bedroom-2-bathroom-0-garage-modern-contemporary-sp324634",
  "https://www.houseplans.com/plan/896-square-feet-1-bedroom-1-5-bathroom-0-garage-contemporary-cottage-beach-cabin-sp268946",
  "https://www.houseplans.com/plan/1276-square-feet-1-bedroom-1-5-bathroom-0-garage-country-ranch-cabin-sp216131",
  "https://www.houseplans.com/plan/1417-square-feet-2-bedroom-2-bathroom-0-garage-cabin-country-cottage-traditional-sp216021",
  "https://www.houseplans.com/plan/991-square-feet-2-bedroom-2-bathroom-0-garage-modern-cabin-ranch-beach-40613",
  "https://www.houseplans.com/plan/1209-square-feet-2-bedroom-1-bathroom-0-garage-modern-contemporary-cabin-sp289227",
  "https://www.houseplans.com/plan/5140-square-feet-5-bedroom-4-50-bathroom-2-garage-log-craftsman-country-cabin-sp123215",
  "https://www.houseplans.com/plan/2348-square-feet-4-bedroom-3-bathroom-2-garage-farmhouse-country-cabin-sp303376",
  "https://www.houseplans.com/plan/1370-square-feet-3-bedroom-2-00-bathroom-0-garage-cabin-country-traditional-sp218070",
  "https://www.houseplans.com/plan/1186-square-feet-2-bedroom-2-bathroom-1-garage-country-sp319291",
  "https://www.houseplans.com/plan/2281-square-feet-3-bedrooms-3-bathroom-coastal-home-plans-0-garage-27613",
  "https://www.houseplans.com/plan/2384-square-feet-3-bedroom-3-5-bathroom-1-garage-farmhouse-modern-sp303353",
  "https://www.houseplans.com/plan/1154-square-feet-2-bedrooms-2-bathroom-northwest-house-plans-0-garage-15056",
  "https://www.houseplans.com/plan/1568-square-feet-2-bedroom-2-5-bathroom-2-garage-log-cabin-39120",
  "https://www.houseplans.com/plan/2145-square-feet-3-bedrooms-2-5-bathroom-waterfront-home-plans-2-garage-27750",
  "https://www.houseplans.com/plan/598-square-feet-1-bedrooms-1-bathroom-beach-home-plans-0-garage-33160",
  "https://www.houseplans.com/plan/761-square-feet-2-bedrooms-1-bathroom-cottage-house-plans-0-garage-4473",
  "https://www.houseplans.com/plan/686-square-feet-2-bedroom-1-00-bathroom-0-garage-contemporary-cabin-country-ranch-sp215967",
  "https://www.houseplans.com/plan/682-square-feet-2-bedroom-1-bathroom-1-garage-modern-cabin-contemporary-sp299640",
  "https://www.houseplans.com/plan/1512-square-feet-3-bedroom-2-5-bathroom-0-garage-cottage-contemporary-farmhouse-cabin-sp289215",
  "https://www.houseplans.com/plan/1605-square-feet-3-bedroom-2-bathroom-0-garage-country-craftsman-cabin-southern-sp301696",
  "https://www.houseplans.com/plan/1286-square-feet-3-bedroom-2-00-bathroom-0-garage-cabin-european-country-sp124339",
  "https://www.houseplans.com/plan/2401-square-feet-1-bedroom-2-bathroom-1-garage-cabin-bungalow-39104",
  "https://www.houseplans.com/plan/5472-square-feet-5-bedroom-4-5-bathroom-1-garage-contemporary-cabin-modern-sp217975",
  "https://www.houseplans.com/plan/1844-square-feet-2-bedroom-2-bathroom-1-garage-cabin-sp273947",
  "https://www.houseplans.com/plan/1697-square-feet-3-bedroom-2-bathroom-0-garage-country-contemporary-beach-cabin-sp270838",
  "https://www.houseplans.com/plan/1050-square-feet-3-bedroom-2-bathroom-0-garage-traditional-cabin-sp325184",
  "https://www.houseplans.com/plan/576-square-feet-1-bedroom-1-bathroom-0-garage-craftsman-39858",
  "https://www.houseplans.com/plan/582-square-feet-1-bedrooms-1-bathroom-cottage-house-plans-0-garage-18420",
  "https://www.houseplans.com/plan/1340-square-feet-3-bedroom-2-bathroom-0-garage-country-39979",
  "https://www.houseplans.com/plan/640-square-feet-1-bedrooms-1-bathroom-contemporary-house-plans-0-garage-36684",
  "https://www.houseplans.com/plan/2346-square-feet-2-bedrooms-3-1-bathroom-cape-cod-house-plans-2-garage-33728",
  "https://www.houseplans.com/plan/950-square-feet-1-bedrooms-1-bathroom-country-house-plans-0-garage-31850",
  "https://www.houseplans.com/plan/1024-square-feet-2-bedrooms-1-bathroom-modern-house-plan-2-garage-35668",
  "https://www.houseplans.com/plan/1240-square-feet-1-bedroom-1-bathroom-1-garage-contemporary-cabin-craftsman-40576",
  "https://www.houseplans.com/plan/808-square-feet-1-bedroom-1-bathroom-0-garage-bungalow-cabin-craftsman-cottage-sp271903",
  "https://www.houseplans.com/plan/1113-square-feet-2-bedroom-1-bathroom-0-garage-european-cabin-craftsman-bungalow-sp123142",
  "https://www.houseplans.com/plan/900-square-feet-2-bedroom-1-bathroom-0-garage-ranch-cabin-country-farmhouse-sp257202",
  "https://www.houseplans.com/plan/440-square-feet-1-bedroom-1-bathroom-0-garage-cabin-ranch-modern-farmhouse-41111",
  "https://www.houseplans.com/plan/1561-square-feet-3-bedroom-2-bathroom-2-garage-craftsman-beach-traditional-cabin-sp268931",
  "https://www.houseplans.com/plan/1235-square-feet-3-bedrooms-2-bathroom-contemporary-house-plans-0-garage-15055",
  "https://www.houseplans.com/plan/1538-square-feet-3-bedroom-2-bathroom-0-garage-cabin-39012",
  "https://www.houseplans.com/plan/3304-square-feet-2-bedrooms-3-bathroom-coastal-home-plans-2-garage-27582",
  "https://www.houseplans.com/plan/1545-square-feet-2-bedroom-2-bathroom-2-garage-craftsman-cabin-ranch-40787",
  "https://www.houseplans.com/plan/3184-square-feet-3-bedrooms-3-bathroom-northwest-house-plans-0-garage-33154",
  "https://www.houseplans.com/plan/2141-square-feet-3-bedrooms-2-bathroom-northwest-house-plans-0-garage-33155",
  "https://www.houseplans.com/plan/900-square-feet-2-bedrooms-1-bathroom-farm-house-plans-0-garage-986",
  "https://www.houseplans.com/plan/682-square-feet-2-bedroom-1-bathroom-0-garage-cottage-sp269661",
  "https://www.houseplans.com/plan/1704-square-feet-3-bedrooms-2-bathroom-beach-home-plans-2-garage-36822",
  "https://www.houseplans.com/plan/1405-square-feet-3-bedroom-2-00-bathroom-0-garage-sp184810",
  "https://www.houseplans.com/plan/704-square-feet-1-bedrooms-1-bathroom-country-house-plans-0-garage-36033",
  "https://www.houseplans.com/plan/2197-square-feet-3-bedroom-2-bathroom-2-garage-cabin-39072",
  "https://www.houseplans.com/plan/1937-square-feet-4-bedrooms-2-5-bathroom-northwest-house-plans-1-garage-36607",
  "https://www.houseplans.com/plan/1170-square-feet-2-bedroom-1-bathroom-0-garage-modern-european-cottage-cabin-sp274421",
  "https://www.houseplans.com/plan/1117-square-feet-2-bedrooms-1-bathroom-northwest-house-plans-0-garage-35506",
  "https://www.houseplans.com/plan/1711-square-feet-3-bedrooms-2-bathroom-southern-house-plans-0-garage-37044",
  "https://www.houseplans.com/plan/2456-square-feet-4-bedrooms-3-bathroom-french-country-plans-2-garage-33966",
  "https://www.houseplans.com/plan/1375-square-feet-2-bedrooms-2-bathroom-traditional-house-plans-0-garage-30488",
  "https://www.houseplans.com/plan/696-square-feet-2-bedroom-1-bathroom-0-garage-craftsman-cabin-country-sp291408",
  "https://www.houseplans.com/plan/1329-square-feet-2-bedroom-3-bathroom-0-garage-craftsman-40861",
  "https://www.houseplans.com/plan/1773-square-feet-3-bedroom-2-5-bathroom-2-garage-country-cabin-southern-traditional-sp297847",
  "https://www.houseplans.com/plan/1338-square-feet-3-bedroom-2-00-bathroom-0-garage-country-sp94212",
  "https://www.houseplans.com/plan/2010-square-feet-4-bedroom-2-bathroom-0-garage-cabin-farmhouse-sp318545",
  "https://www.houseplans.com/plan/1124-square-feet-2-bedroom-2-bathroom-0-garage-cabin-country-farmhouse-cottage-sp348145",
  "https://www.houseplans.com/plan/949-square-feet-1-bedroom-1-bathroom-0-garage-cabin-cottage-sp323456",
  "https://www.houseplans.com/plan/3398-square-feet-4-bedroom-3-bathroom-1-garage-contemporary-log-cabin-sp257057",
  "https://www.houseplans.com/plan/2133-square-feet-3-bedroom-2-bathroom-0-garage-contemporary-cabin-ranch-craftsman-sp317978",
  "https://www.houseplans.com/plan/2695-square-feet-3-bedroom-2-5-bathroom-2-garage-ranch-cabin-craftsman-farmhouse-sp322124",
  "https://www.houseplans.com/plan/6218-square-feet-5-bedroom-3-5-bathroom-5-garage-country-sp327378cabin-cottage-modern-sp317785",
  "https://www.houseplans.com/plan/1906-square-feet-3-bedroom-2-5-bathroom-2-garage-country-cottage-cabin-sp289241",
  "https://www.houseplans.com/plan/1800-square-feet-2-bedroom-2-5-bathroom-2-garage-cottage-cabin-farmhouse-sp291961",
  "https://www.houseplans.com/plan/1557-square-feet-3-bedrooms-2-bathroom-contemporary-house-plans-0-garage-26261",
  "https://www.houseplans.com/plan/2760-square-feet-4-bedroom-3-5-bathroom-0-garage-cabin-sp269214",
  "https://www.houseplans.com/plan/780-square-feet-2-bedroom-1-bathroom-0-garage-cabin-cottage-ranch-modern-41202",
  "https://www.houseplans.com/plan/681-square-feet-2-bedroom-2-00-bathroom-0-garage-cabin-log-country-ranch-sp300305",
  "https://www.houseplans.com/plan/1071-square-feet-2-bedroom-1-5-bathroom-0-garage-cabin-craftsman-southern-traditional-sp303250",
  "https://www.houseplans.com/plan/1039-square-feet-2-bedroom-2-bathroom-0-garage-classical-cabin-country-sp325799",
  "https://www.houseplans.com/plan/2697-square-feet-2-bedroom-2-bathroom-0-garage-cabin-sp265304",
  "https://www.houseplans.com/plan/689-square-feet-2-bedroom-1-bathroom-0-garage-craftsman-cottage-cabin-bungalow-sp316485",
  "https://www.houseplans.com/plan/970-square-feet-3-bedroom-1-bathroom-0-garage-country-cabin-traditional-sp216105",
  "https://www.houseplans.com/plan/3307-square-feet-3-bedroom-4-5-bathroom-0-garage-cabin-craftsman-40863",
  "https://www.houseplans.com/plan/2587-square-feet-3-bedroom-2-5-bathroom-3-garage-craftsman-sp298609",
  "https://www.houseplans.com/plan/791-square-feet-1-bedroom-2-bathroom-0-garage-traditional-bungalow-cabin-cottage-sp257805",
  "https://www.houseplans.com/plan/1380-square-feet-2-bedrooms-2-bathroom-contemporary-house-plans-2-garage-35182",
  "https://www.houseplans.com/plan/1800-square-feet-3-bedroom-2-bathroom-2-garage-ranch-farmhouse-country-cabin-sp347382",
  "https://www.houseplans.com/plan/1140-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp325825",
  "https://www.houseplans.com/plan/1815-square-feet-3-bedroom-2-bathroom-2-garage-country-cabin-craftsman-farmhouse-sp347122",
  "https://www.houseplans.com/plan/2926-square-feet-4-bedroom-3-bathroom-1-garage-country-cabin-sp302258",
  "https://www.houseplans.com/plan/4720-square-feet-5-bedroom-4-bathroom-2-garage-cabin-cottage-contemporary-sp324407",
  "https://www.houseplans.com/plan/12163-square-feet-0-bedroom-3-bathroom-6-garage-farmhouse-traditional-cabin-sp271635",
  "https://www.houseplans.com/plan/2106-square-feet-4-bedroom-2-5-bathroom-0-garage-cabin-farmhouse-country-sp321887",
  "https://www.houseplans.com/plan/1271-square-feet-1-bedroom-1-bathroom-3-garage-farmhouse-cabin-craftsman-sp319324",
  "https://www.houseplans.com/plan/1972-square-feet-0-bedroom-0-bathroom-4-garage-european-traditional-cabin-sp314743",
  "https://www.houseplans.com/plan/1268-square-feet-0-bedroom-0-bathroom-3-garage-traditional-country-cottage-cabin-sp291957",
  "https://www.houseplans.com/plan/4448-square-feet-1-bedroom-2-bathroom-2-garage-craftsman-ranch-country-cabin-sp291432",
  "https://www.houseplans.com/plan/1379-square-feet-3-bedrooms-1-bathroom-cottage-house-plans-0-garage-16486",
  "https://www.houseplans.com/plan/2000-square-feet-3-bedroom-2-bathroom-0-garage-cabin-sp325497",
  "https://www.houseplans.com/plan/1333-square-feet-2-bedroom-2-bathroom-0-garage-cabin-sp314750",
  "https://www.houseplans.com/plan/725-square-feet-2-bedroom-1-bathroom-0-garage-cottage-craftsman-cabin-bungalow-sp325289",
  "https://www.houseplans.com/plan/860-square-feet-1-bedrooms-1-bathroom-contemporary-house-plans-2-garage-36567",
  "https://www.houseplans.com/plan/2030-square-feet-4-bedroom-2-bathroom-0-garage-cabin-farmhouse-sp318540",
  "https://www.houseplans.com/plan/5828-square-feet-6-bedrooms-4-5-bathroom-luxury-home-plans-4-garage-36317",
  "https://www.houseplans.com/plan/2854-square-feet-4-bedroom-3-5-bathroom-2-garage-country-craftsman-farmhouse-cabin-sp346359",
  "https://www.houseplans.com/plan/2897-square-feet-4-bedroom-3-5-bathroom-2-garage-modern-country-cabin-ranch-sp346352",
  "https://www.houseplans.com/plan/2524-square-feet-4-bedroom-2-5-bathroom-2-garage-modern-farmhouse-country-cabin-sp346749",
  "https://www.houseplans.com/plan/1198-square-feet-2-bedroom-2-bathroom-0-garage-craftsman-cottage-cabin-sp331887",
  "https://www.houseplans.com/plan/1968-square-feet-2-bedrooms-2-bathroom-country-house-plans-2-garage-36341",
  "https://www.houseplans.com/plan/2295-square-feet-2-bedroom-2-bathroom-0-garage-contemporary-modern-cabin-sp217848",
  "https://www.houseplans.com/plan/2394-square-feet-3-bedroom-2-5-bathroom-0-garage-craftsman-country-cabin-sp303306",
  "https://www.houseplans.com/plan/3806-square-feet-4-bedroom-4-bathroom-0-garage-cabin-sp303300",
  "https://www.houseplans.com/plan/993-square-feet-2-bedroom-2-bathroom-2-garage-farmhouse-country-cabin-sp303332",
  "https://www.houseplans.com/plan/2699-square-feet-3-bedroom-2-bathroom-1-garage-farmhouse-cottage-craftsman-cabin-sp303333",
  "https://www.houseplans.com/plan/4189-square-feet-3-bedroom-2-5-bathroom-1-garage-craftsman-european-cabin-sp303319",
  "https://www.houseplans.com/plan/2884-square-feet-3-bedroom-3-5-bathroom-1-garage-farmhouse-craftsman-cabin-sp303334",
  "https://www.houseplans.com/plan/3005-square-feet-3-bedroom-2-5-bathroom-0-garage-cabin-cottage-craftsman-sp303336",
  "https://www.houseplans.com/plan/5309-square-feet-4-bedroom-3-5-bathroom-1-garage-farmhouse-country-craftsman-cabin-sp303339",
  "https://www.houseplans.com/plan/3820-square-feet-3-bedrooms-2-5-bathroom-coastal-home-plans-2-garage-27589",
  "https://www.houseplans.com/plan/6609-square-feet-3-bedroom-4-bathroom-1-garage-contemporary-adobe-southwestern-sp319299",
  "https://www.houseplans.com/plan/4285-square-feet-3-bedroom-4-bathroom-1-garage-craftsman-traditional-sp303346",
  "https://www.houseplans.com/plan/3328-square-feet-4-bedroom-2-5-bathroom-1-garage-farmhouse-country-cabin-sp303343",
  "https://www.houseplans.com/plan/1706-square-feet-2-bedroom-2-bathroom-0-garage-traditional-cabin-cottage-sp222285",
  "https://www.houseplans.com/plan/1360-square-feet-2-bedroom-2-bathroom-0-garage-bungalow-cabin-cottage-sp324095",
  "https://www.houseplans.com/plan/1752-square-feet-2-bedroom-2-5-bathroom-0-garage-country-cabin-cottage-sp336868",
  "https://www.houseplans.com/plan/1072-square-feet-1-bedroom-1-5-bathroom-0-garage-modern-cabin-contemporary-cottage-sp317015",
  "https://www.houseplans.com/plan/982-square-feet-2-bedroom-1-bathroom-0-garage-cottage-cabin-sp296418",
  "https://www.houseplans.com/plan/1487-square-feet-2-bedroom-1-bathroom-0-garage-bungalow-cabin-country-contemporary-sp316324",
  "https://www.houseplans.com/plan/946-square-feet-2-bedrooms-1-bathroom-contemporary-house-plans-0-garage-2017",
  "https://www.houseplans.com/plan/992-square-feet-2-bedroom-1-bathroom-0-garage-country-39777",
  "https://www.houseplans.com/plan/1293-square-feet-3-bedroom-2-bathroom-0-garage-country-39809",
  "https://www.houseplans.com/plan/2638-square-feet-3-bedroom-1-bathroom-1-garage-cabin-40027",
  "https://www.houseplans.com/plan/1960-square-feet-4-bedroom-2-bathroom-0-garage-country-39863",
  "https://www.houseplans.com/plan/860-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp318542",
  "https://www.houseplans.com/plan/1100-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp318543",
  "https://www.houseplans.com/plan/860-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp318542",
  "https://www.houseplans.com/plan/1100-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp318543",
  "https://www.houseplans.com/plan/1592-square-feet-3-bedroom-2-bathroom-0-garage-cabin-sp318549",
  "https://www.houseplans.com/plan/648-square-feet-1-bedroom-1-bathroom-0-garage-farmhouse-cottage-cabin-sp346847",
  "https://www.houseplans.com/plan/730-square-feet-1-bedroom-1-bathroom-0-garage-cottage-cabin-sp333204",
  "https://www.houseplans.com/plan/2683-square-feet-4-bedroom-5-bathroom-2-garage-craftsman-cabin-sp321835",
  "https://www.houseplans.com/plan/2979-square-feet-4-bedroom-4-5-bathroom-2-garage-contemporary-modern-cabin-sp337243",
  "https://www.houseplans.com/plan/2221-square-feet-3-bedroom-2-5-bathroom-1-garage-country-farmhouse-sp301583",
  "https://www.houseplans.com/plan/4493-square-feet-5-bedroom-4-5-bathroom-2-garage-craftsman-cabin-39618",
  "https://www.houseplans.com/plan/1371-square-feet-1-bedroom-1-bathroom-3-garage-cabin-craftsman-sp323686",
  "https://www.houseplans.com/plan/1660-square-feet-3-bedroom-3-bathroom-0-garage-cabin-country-southern-sp323693",
  "https://www.houseplans.com/plan/4254-square-feet-4-bedroom-4-bathroom-3-garage-cottage-traditional-cabin-craftsman-sp336994",
  "https://www.houseplans.com/plan/2961-square-feet-3-bedroom-3-5-bathroom-2-garage-cottage-southern-beach-cabin-sp337009",
  "https://www.houseplans.com/plan/3325-square-feet-4-bedroom-3-5-bathroom-0-garage-country-craftsman-cabin-southern-sp330989",
  "https://www.houseplans.com/plan/1442-square-feet-2-bedroom-2-bathroom-2-garage-ranch-cabin-prairie-southern-sp331981",
  "https://www.houseplans.com/plan/2725-square-feet-3-bedroom-3-bathroom-2-garage-country-farmhouse-cabin-southern-sp333631",
  "https://www.houseplans.com/plan/1343-square-feet-2-bedroom-2-bathroom-0-garage-cabin-sp318548",
  "https://www.houseplans.com/plan/1397-square-feet-3-bedroom-2-bathroom-2-garage-ranch-southern-country-cabin-sp337005",
  "https://www.houseplans.com/plan/845-square-feet-2-bedroom-1-bathroom-0-garage-cabin-sp318546",
  "https://www.houseplans.com/plan/1979-square-feet-3-bedroom-2-bathroom-0-garage-cabin-40833",
  "https://www.houseplans.com/plan/2400-square-feet-3-bedroom-2-5-bathroom-0-garage-southern-modern-contemporary-sp303369",
  "https://www.houseplans.com/plan/1683-square-feet-2-bedroom-2-bathroom-1-garage-farmhouse-country-craftsman-sp345710",
  "https://www.houseplans.com/plan/1465-square-feet-2-bedroom-2-bathroom-3-garage-southern-country-cabin-sp304120",
  "https://www.houseplans.com/plan/1356-square-feet-3-bedroom-2-bathroom-0-garage-modern-39082",
  "https://www.houseplans.com/plan/1550-square-feet-3-bedroom-2-5-bathroom-2-garage-log-farmhouse-cabin-sp315979",
  "https://www.houseplans.com/plan/3458-square-feet-4-bedroom-4-5-bathroom-3-garage-modern-contemporary-craftsman-sp215747",
  "https://www.houseplans.com/plan/1411-square-feet-3-bedroom-2-bathroom-1-garage-contemporary-modern-sp268942",
  "https://www.houseplans.com/plan/2373-square-feet-4-bedroom-2-5-bathroom-2-garage-european-contemporary-southern-sp216857",
  "https://www.houseplans.com/plan/800-square-feet-2-bedroom-1-bathroom-0-garage-modern-37865",
  "https://www.houseplans.com/plan/2374-square-feet-3-bedroom-2-5-bathroom-1-garage-modern-contemporary-sp324415",
  "https://www.houseplans.com/plan/1642-square-feet-3-bedroom-2-5-bathroom-2-garage-farmhouse-modern-sp289231",
  "https://www.houseplans.com/plan/1200-square-feet-2-bedroom-2-bathroom-0-garage-cabin-modern-contemporary-country-sp261046",
  "https://www.houseplans.com/plan/2802-square-feet-4-bedroom-4-bathroom-1-garage-traditional-sp259367",
  "https://www.houseplans.com/plan/3986-square-feet-4-bedroom-3-5-bathroom-2-garage-modern-european-contemporary-beach-sp123694",
  "https://www.houseplans.com/plan/1105-square-feet-2-bedroom-2-bathroom-0-garage-modern-beach-contemporary-sp292454",
  "https://www.houseplans.com/plan/714-square-feet-3-bedroom-2-5-bathroom-0-garage-farmhouse-modern-country-sp267839",
  "https://www.houseplans.com/plan/1600-square-feet-3-bedroom-2-bathroom-2-garage-farmhouse-modern-sp296338",
  "https://www.houseplans.com/plan/1323-square-feet-2-bedroom-2-bathroom-0-garage-contemporary-modern-sp264786",
  "https://www.houseplans.com/plan/3304-square-feet-4-bedroom-3-5-bathroom-1-garage-contemporary-sp300444",
  "https://www.houseplans.com/plan/1800-square-feet-3-bedroom-2-bathroom-0-garage-contemporary-sp324751",
  "https://www.houseplans.com/plan/2557-square-feet-3-bedrooms-2-5-bathroom-contemporary-house-plans-2-garage-35285",
  "https://www.houseplans.com/plan/3346-square-feet-4-bedroom-3-5-bathroom-3-garage-farmhouse-modern-sp314671",
  "https://www.houseplans.com/plan/2150-square-feet-4-bedroom-4-5-bathroom-0-garage-farmhouse-ranch-cabin-southern-sp337177",
  "https://www.houseplans.com/plan/1953-square-feet-2-bedroom-2-5-bathroom-2-garage-modern-39702",
  "https://www.houseplans.com/plan/2287-square-feet-3-bedroom-3-bathroom-2-garage-contemporary-modern-sp217183",
  "https://www.houseplans.com/plan/2621-square-feet-4-bedroom-4-5-bathroom-2-garage-adobe-southwestern-southern-contemporary-modern-sp284942",
  "https://www.houseplans.com/plan/3310-square-feet-4-bedroom-3-5-bathroom-2-garage-modern-contemporary-beach-european-sp123506",
  "https://www.houseplans.com/plan/1067-square-feet-1-bedroom-1-bathroom-0-garage-contemporary-modern-sp328747",
  "https://www.houseplans.com/plan/3374-square-feet-3-bedroom-3-5-bathroom-0-garage-farmhouse-bungalow-39288",
  "https://www.houseplans.com/plan/850-square-feet-2-bedroom-1-bathroom-0-garage-modern-cabin-40508",
  "https://www.houseplans.com/plan/2142-square-feet-4-bedroom-3-bathroom-2-garage-modern-contemporary-40886",
  "https://www.houseplans.com/plan/3482-square-feet-5-bedroom-4-5-bathroom-3-garage-farmhouse-modern-sp321878",
  "https://www.houseplans.com/plan/1633-square-feet-3-bedrooms-2-bathroom-sunbelt-home-plans-2-garage-36400",
  "https://www.houseplans.com/plan/1484-square-feet-3-bedroom-2-bathroom-1-garage-cabin-sp318541",
  "https://www.houseplans.com/plan/1912-square-feet-3-bedroom-2-5-bathroom-1-garage-contemporary-modern-farmhouse-sp296274",
  "https://www.houseplans.com/plan/2042-square-feet-3-bedroom-2-5-bathroom-2-garage-contemporary-modern-sp214871",
  "https://www.houseplans.com/plan/926-square-feet-2-bedroom-2-bathroom-1-garage-modern-cottage-ranch-bungalow-sp315778",
  "https://www.houseplans.com/plan/3334-square-feet-4-bedroom-3-50-bathroom-2-garage-contemporary-prairie-modern-sp154713",
  "https://www.houseplans.com/plan/5293-square-feet-5-bedroom-4-5-bathroom-3-garage-contemporary-modern-sp289608",
  "https://www.houseplans.com/plan/1162-square-feet-2-bedroom-2-bathroom-0-garage-modern-european-contemporary-cabin-sp292670",
  "https://www.houseplans.com/plan/1655-square-feet-2-bedroom-2-5-bathroom-0-garage-modern-sp325812",
  "https://www.houseplans.com/plan/860-square-feet-2-bedrooms-2-bathroom-modern-house-plan-0-garage-36670",
  "https://www.houseplans.com/plan/1188-square-feet-2-bedroom-2-bathroom-1-garage-modern-contemporary-cabin-cottage-sp264246",
  "https://www.houseplans.com/plan/6136-square-feet-5-bedroom-5-1-bathroom-4-garage-contemporary-european-mediterranean-sp94277",
  "https://www.houseplans.com/plan/2370-square-feet-3-bedroom-2-bathroom-2-garage-contemporary-39865",
  "https://www.houseplans.com/plan/4983-square-feet-4-bedroom-3-50-bathroom-3-garage-contemporary-traditional-sp124306",
  "https://www.houseplans.com/plan/1913-square-feet-3-bedroom-2-bathroom-2-garage-modern-contemporary-sp303160",
  "https://www.houseplans.com/plan/3828-square-feet-4-bedroom-2-5-bathroom-3-garage-farmhouse-sp273892",
];


# Directory to save images
image_dir_base = 'houseplan_images'
if not os.path.exists(image_dir_base):
    os.makedirs(image_dir_base)

# List to hold all scraped data
all_scraped_data = []

# Function to scrape data from a single URL
def scrape_data(url, index):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    divelemet = soup.find_all('img', class_="w-full", height=533)
    labels = soup.find_all('div', class_='lg:max-w-601px mb-5 pt-2 px-3 lg:px-0 order-2 sm:order-1')

    image_link_data = []
    category_dict = {}

    for elem in divelemet:
        image_link = elem.get('data-lazy', '')
        if image_link:
            image_link_data.append(image_link)

    for texts in labels:
        x = texts.find_all('div', class_="block")
        for it in x:
            d = it.find_all('h3')
            for every_title in d:
                category_title = every_title.get_text().strip()
                if category_title not in category_dict:
                    category_dict[category_title] = []
            sub = it.find_all('div', class_="flex")
            for text_data in sub:
                category_text = text_data.find_all('div', class_="w-1/2 text-xs md:text-sm font-normal leading-normal")
                for each_item in category_text:
                    category_value = each_item.get_text().strip()
                    if category_title:
                        category_dict[category_title].append(category_value)

    image_path_data = []
    # Create a directory for the URL
    url_dir = os.path.join(image_dir_base, f'url_{index + 1}')
    if not os.path.exists(url_dir):
        os.makedirs(url_dir)

    for img_index, image_url in enumerate(image_link_data):
        if image_url:
            image_response = requests.get(image_url)
            img_path = os.path.join(url_dir, f'image_{img_index + 1}.jpg')
            with open(img_path, 'wb') as file:
                file.write(image_response.content)
            image_path_data.append(img_path)

    return {
        "url": url,
        "image_url": image_path_data,
        "categories": category_dict
    }

# Scrape data from each URL and append to the list
for index, url in enumerate(tqdm(urls, desc="Scraping URLs")):
    scraped_data = scrape_data(url, index)
    all_scraped_data.append(scraped_data)

# Save all scraped data to a JSON file
with open('all_scraped_data.json', 'w') as json_file:
    json.dump(all_scraped_data, json_file, indent=4)

# Print all scraped data
print(json.dumps(all_scraped_data, indent=4))
```
